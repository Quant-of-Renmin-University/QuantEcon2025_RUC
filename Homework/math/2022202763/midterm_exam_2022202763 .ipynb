```python
# äºŒã€æ•°æ®é¢„å¤„ç†
# 2.11 test_ç¯çº¿
def preprocess_ring(df):
    """
    ç¯çº¿åˆ—é¢„å¤„ç†å‡½æ•°ï¼ˆé€‚ç”¨äºtrainå’Œtestæ•°æ®é›†ï¼‰
    å¤„ç†é€»è¾‘ï¼š
    1. å¡«è¡¥ç©ºç™½å€¼ä¸º"å¤–ç¯å¤–"ï¼ˆæœ€æ™®éæƒ…å†µï¼‰
    2. ç»Ÿä¸€ä¸åŒè¡¨è¾¾æ–¹å¼
    3. æœ‰åºç¼–ç ï¼ˆæ•°å€¼è¶Šå¤§è¡¨ç¤ºç¦»å¸‚ä¸­å¿ƒè¶Šè¿œï¼‰
    """
    # å¤åˆ¶æ•°æ®é¿å…ä¿®æ”¹åŸDataFrame
    df = df.copy()
    
    # ç»Ÿä¸€è¡¨è¾¾æ–¹å¼
    ring_mapping = {
        'å†…ç¯å†…': 'ä¸€ç¯å†…',
        'äºŒç¯å†…': 'ä¸€ç¯è‡³äºŒç¯',
        'å†…ç¯è‡³ä¸­ç¯': 'ä¸€ç¯è‡³äºŒç¯',
        'å†…ç¯è‡³å¤–ç¯': 'ä¸€ç¯è‡³å¤–ç¯',
        'ä¸­ç¯è‡³å¤–ç¯': 'äºŒç¯è‡³å¤–ç¯',
        'ä¸€è‡³äºŒç¯': 'ä¸€ç¯è‡³äºŒç¯'  # æ–°å¢æ˜ å°„
    }
    
    # å¡«è¡¥ç©ºç™½
    df['ç¯çº¿'] = df['ç¯çº¿'].fillna('å¤–ç¯å¤–')
    
    # ç»Ÿä¸€è¡¨è¾¾
    df['ç¯çº¿'] = df['ç¯çº¿'].replace(ring_mapping)
    
    # æœ‰åºç¼–ç å­—å…¸ï¼ˆæ•°å€¼è¶Šå¤§ç¦»å¸‚ä¸­å¿ƒè¶Šè¿œï¼‰
    ring_order = {
        'ä¸€ç¯å†…': 1,
        'ä¸€ç¯è‡³äºŒç¯': 2,
        'äºŒè‡³ä¸‰ç¯': 3,
        'ä¸‰è‡³å››ç¯': 4,
        'å››è‡³äº”ç¯': 5,
        'äº”è‡³å…­ç¯': 6,
        'ä¸‰ç¯å¤–': 4.5,
        'å››ç¯å¤–': 5.5,
        'å…­ç¯å¤–': 7,
        'ä¸€ç¯è‡³å¤–ç¯': 3.5,
        'äºŒç¯è‡³å¤–ç¯': 4,
        'å¤–ç¯å¤–': 6.5
    }
    
    # åº”ç”¨ç¼–ç 
    df['ç¯çº¿_ç¼–ç '] = df['ç¯çº¿'].map(ring_order)
    
    # æ·»åŠ æ˜¯å¦æ ¸å¿ƒåŒºæ ‡å¿—
    df['æ˜¯å¦æ ¸å¿ƒåŒº'] = df['ç¯çº¿_ç¼–ç '].apply(lambda x: 1 if x <= 3 else 0)
    
    # è¿”å›å¤„ç†åçš„æ•°æ®å’Œç¼–ç å­—å…¸
    return df, ring_order

# ==================== æµ‹è¯•éªŒè¯ ==================== 
# åœ¨è®­ç»ƒé›†ä¸Šåº”ç”¨
train_df, ring_order = preprocess_ring(train_df)

# éªŒè¯å¤„ç†ç»“æœ
print("===== ç¯çº¿å€¼åˆ†å¸ƒ =====")
print(train_df['ç¯çº¿'].value_counts(dropna=False))

print("\n===== ç¼–ç ååˆ†å¸ƒ =====")
print(train_df.groupby('ç¯çº¿')['ç¯çº¿_ç¼–ç '].first())

print("\n===== æ ·æœ¬ç¤ºä¾‹ =====")
print(train_df[['ç¯çº¿', 'ç¯çº¿_ç¼–ç ', 'æ˜¯å¦æ ¸å¿ƒåŒº']].head(10))

# ä¿å­˜ç¼–ç æ˜ å°„ä¾›æµ‹è¯•é›†ä½¿ç”¨
ring_encoding_map = {
    'åŸå§‹å€¼': list(ring_order.keys()),
    'ç¼–ç å€¼': list(ring_order.values())
}
pd.DataFrame(ring_encoding_map).to_csv(f"{output_dir}/ring_encoding_map.csv", index=False)
    
```

```python
# 2.12æµ‹è¯•é›†ç¯çº¿åˆ—é¢„å¤„ç†
def preprocess_test_ring(test_df, encoding_map_path=f"{output_dir}/ring_encoding_map.csv"):
    """
    æµ‹è¯•é›†ç¯çº¿åˆ—é¢„å¤„ç†ï¼ˆä½¿ç”¨è®­ç»ƒé›†ç”Ÿæˆçš„ç¼–ç è§„åˆ™ï¼‰
    å‚æ•°:
        test_df: æµ‹è¯•é›†DataFrame
        encoding_map_path: è®­ç»ƒé›†ä¿å­˜çš„ç¼–ç æ˜ å°„æ–‡ä»¶è·¯å¾„
    """
    # åŠ è½½è®­ç»ƒé›†ç”Ÿæˆçš„ç¼–ç æ˜ å°„
    try:
        ring_encoding = pd.read_csv(encoding_map_path)
        ring_order = dict(zip(ring_encoding['åŸå§‹å€¼'], ring_encoding['ç¼–ç å€¼']))
    except Exception as e:
        print(f"âŒ åŠ è½½ç¼–ç æ˜ å°„å¤±è´¥: {e}")
        return test_df
    
    # å¤åˆ¶æ•°æ®é¿å…ä¿®æ”¹åŸDataFrame
    test_df = test_df.copy()
    
    # ç»Ÿä¸€è¡¨è¾¾æ–¹å¼ï¼ˆä¸è®­ç»ƒé›†ç›¸åŒï¼‰
    ring_mapping = {
        'å†…ç¯å†…': 'ä¸€ç¯å†…',
        'äºŒç¯å†…': 'ä¸€ç¯è‡³äºŒç¯',
        'å†…ç¯è‡³ä¸­ç¯': 'ä¸€ç¯è‡³äºŒç¯',
        'å†…ç¯è‡³å¤–ç¯': 'ä¸€ç¯è‡³å¤–ç¯',
        'ä¸­ç¯è‡³å¤–ç¯': 'äºŒç¯è‡³å¤–ç¯',
        'ä¸€è‡³äºŒç¯': 'ä¸€ç¯è‡³äºŒç¯'
    }
    
    # å¡«è¡¥ç©ºç™½ï¼ˆä¸è®­ç»ƒé›†ç›¸åŒï¼‰
    test_df['ç¯çº¿'] = test_df['ç¯çº¿'].fillna('å¤–ç¯å¤–')
    
    # ç»Ÿä¸€è¡¨è¾¾
    test_df['ç¯çº¿'] = test_df['ç¯çº¿'].replace(ring_mapping)
    
    # åº”ç”¨è®­ç»ƒé›†çš„ç¼–ç è§„åˆ™
    test_df['ç¯çº¿_ç¼–ç '] = test_df['ç¯çº¿'].map(ring_order)
    
    # æ·»åŠ æ˜¯å¦æ ¸å¿ƒåŒºæ ‡å¿—ï¼ˆä¸è®­ç»ƒé›†ç›¸åŒé˜ˆå€¼ï¼‰
    test_df['æ˜¯å¦æ ¸å¿ƒåŒº'] = test_df['ç¯çº¿_ç¼–ç '].apply(lambda x: 1 if x <= 3 else 0)
    
    # æ£€æŸ¥æ˜¯å¦æœ‰æ–°å‡ºç°çš„ç¯çº¿å€¼
    unseen_rings = set(test_df['ç¯çº¿'].unique()) - set(ring_order.keys())
    if unseen_rings:
        print(f"âš ï¸ å‘ç°è®­ç»ƒé›†æœªå‡ºç°çš„ç¯çº¿å€¼: {unseen_rings}")
        print("å°†è®¾ä¸ºé»˜è®¤å€¼'å¤–ç¯å¤–'çš„ç¼–ç ")
        test_df['ç¯çº¿_ç¼–ç '] = test_df['ç¯çº¿_ç¼–ç '].fillna(ring_order['å¤–ç¯å¤–'])
    
    return test_df

# åº”ç”¨é¢„å¤„ç†
test_df = preprocess_test_ring(test_df)

# éªŒè¯å¤„ç†ç»“æœ
print("\n===== æµ‹è¯•é›†ç¯çº¿å¤„ç†éªŒè¯ =====")
print("1. ç¼–ç åç¼ºå¤±å€¼æ•°é‡:", test_df['ç¯çº¿_ç¼–ç '].isnull().sum())
print("2. ç¯çº¿å€¼åˆ†å¸ƒ:")
print(test_df['ç¯çº¿'].value_counts(dropna=False))
print("\n3. æ ·æœ¬ç¤ºä¾‹:")
print(test_df[['ç¯çº¿', 'ç¯çº¿_ç¼–ç ', 'æ˜¯å¦æ ¸å¿ƒåŒº']].head(10))

# ä¿å­˜å¤„ç†åçš„æµ‹è¯•é›†
test_df.to_csv(f"{output_dir}/test_preprocessed.csv", index=False, encoding="utf_8_sig")
print(f"\nğŸ’¾ é¢„å¤„ç†åçš„æµ‹è¯•é›†å·²ä¿å­˜åˆ°: {output_dir}/test_preprocessed.csv")
```

```python
# 2.2 â€œå°åŒºåç§°â€
# åˆ é™¤å°åŒºåç§°åˆ—çš„æ ‡å‡†åŒ–å¤„ç†
def remove_neighborhood_col(df, col_name='å°åŒºåç§°'):
    """
    åˆ é™¤æŒ‡å®šåˆ—å¹¶è®°å½•æ“ä½œæ—¥å¿—
    å‚æ•°:
        df: è¾“å…¥DataFrame
        col_name: è¦åˆ é™¤çš„åˆ—å
    è¿”å›:
        å¤„ç†åçš„DataFrame
    """
    if col_name in df.columns:
        print(f"âœ… æ­£åœ¨åˆ é™¤åˆ—: {col_name}")
        df = df.drop(columns=[col_name])
        # è®°å½•åˆ é™¤æ“ä½œ
        with open(f"{output_dir}/data_preprocessing_log.txt", 'a') as f:
            f.write(f"[{pd.Timestamp.now()}] åˆ é™¤åˆ—: {col_name}\n")
    else:
        print(f"âš ï¸ åˆ— {col_name} ä¸å­˜åœ¨ï¼Œè·³è¿‡åˆ é™¤")
    return df

# åº”ç”¨åˆ é™¤æ“ä½œ
print("===== è®­ç»ƒé›†å¤„ç† =====")
train_df = remove_neighborhood_col(train_df)

print("\n===== æµ‹è¯•é›†å¤„ç† =====")
test_df = remove_neighborhood_col(test_df)

# éªŒè¯åˆ é™¤ç»“æœ
print("\n===== å¤„ç†ç»“æœéªŒè¯ =====")
print("è®­ç»ƒé›†å‰©ä½™åˆ—:", train_df.columns.tolist())
print("æµ‹è¯•é›†å‰©ä½™åˆ—:", test_df.columns.tolist())

# ä¿å­˜å¤„ç†åçš„æ•°æ®
train_df.to_csv(f"{output_dir}/train_no_neighborhood.csv", index=False, encoding='utf_8_sig')
test_df.to_csv(f"{output_dir}/test_no_neighborhood.csv", index=False, encoding='utf_8_sig')

print(f"\nğŸ’¾ å¤„ç†åçš„æ•°æ®å·²ä¿å­˜:")
print(f"- {output_dir}/train_no_neighborhood.csv")
print(f"- {output_dir}/test_no_neighborhood.csv")
print(f"- æ“ä½œæ—¥å¿—: {output_dir}/data_preprocessing_log.txt")
```

```python
# 2.31æˆ¿å±‹æˆ·å‹test



import os
import pandas as pd
from functools import reduce

# ç¡®ä¿æ‰€æœ‰ç›®å½•å­˜åœ¨
output_dir = "/home/mw/project/output"
columns_dir = os.path.join(output_dir, "columns")
os.makedirs(columns_dir, exist_ok=True)  # å…³é”®ä¿®å¤ï¼šåˆ›å»ºcolumnså­ç›®å½•

def preprocess_house_layout(df, is_train=True):
    """æˆ¿å±‹æˆ·å‹åˆ—é¢„å¤„ç†å‡½æ•°ï¼ˆä¸ä¹‹å‰ä¿æŒä¸€è‡´ï¼‰"""
    df = df.copy()
    df['æˆ¿å±‹æˆ·å‹'] = df['æˆ¿å±‹æˆ·å‹'].str.replace('æˆ¿é—´', 'å®¤').str.replace('å«å®¤', 'å« å®¤')
    df['æˆ¿å±‹æˆ·å‹'] = df['æˆ¿å±‹æˆ·å‹'].fillna('1å®¤1å…1å¨1å«')
    
    pattern = r'(\d+)å®¤(\d*)å…(\d*)å¨(\d*)å«'
    extracted = df['æˆ¿å±‹æˆ·å‹'].str.extract(pattern)
    
    result = pd.DataFrame()
    result['å§å®¤æ•°'] = extracted[0].fillna('0').astype(int)
    result['å®¢å…æ•°'] = extracted[1].fillna('0').astype(int)
    result['å¨æˆ¿æ•°'] = extracted[2].fillna('0').astype(int)
    result['å«ç”Ÿé—´æ•°'] = extracted[3].fillna('0').astype(int)
    
    result.loc[result['å§å®¤æ•°'] > 10, 'å§å®¤æ•°'] = 10
    result.loc[result['å«ç”Ÿé—´æ•°'] > result['å§å®¤æ•°'] + 3, 'å«ç”Ÿé—´æ•°'] = result['å§å®¤æ•°'] + 3
    
    result['æ€»æˆ¿é—´æ•°'] = result['å§å®¤æ•°'] + result['å®¢å…æ•°']
    result['å§å®¤å«ç”Ÿé—´æ¯”'] = result['å§å®¤æ•°'] / (result['å«ç”Ÿé—´æ•°'] + 0.01)
    result['æ˜¯å¦æ ‡å‡†æˆ·å‹'] = ((result['å§å®¤æ•°'] >= 1) & 
                          (result['å®¢å…æ•°'] >= 1) & 
                          (result['å¨æˆ¿æ•°'] >= 1) & 
                          (result['å«ç”Ÿé—´æ•°'] >= 1)).astype(int)
    result['æ˜¯å¦é›¶å±…å®¤'] = (result['å§å®¤æ•°'] == 0).astype(int)
    result['æ˜¯å¦è±ªå®…æˆ·å‹'] = (result['å§å®¤æ•°'] >= 4).astype(int)
    
    result = result.add_prefix('æˆ·å‹_')
    return result, df['æˆ¿å±‹æˆ·å‹']

# ==================== è®­ç»ƒé›†å¤„ç† ====================
try:
    train_path = os.path.join(output_dir, "train_stage1_preprocessed.csv")
    train_df = pd.read_csv(train_path)
    
    # åº”ç”¨æˆ·å‹é¢„å¤„ç†
    train_layout_features, train_original_layout = preprocess_house_layout(train_df)
    
    # ä¿å­˜æˆ·å‹ç‰¹å¾ï¼ˆä¿®æ­£è·¯å¾„ï¼‰
    train_layout_path = os.path.join(columns_dir, "train_æˆ¿å±‹æˆ·å‹.csv")
    train_layout_features.to_csv(train_layout_path, index=False, encoding='utf_8_sig')
    print(f"âœ… æˆ·å‹ç‰¹å¾å·²ä¿å­˜åˆ°: {train_layout_path}")

    # æ›´æ–°åˆå¹¶æ–‡ä»¶
    def update_merged_file():
        merged_train = pd.read_csv(os.path.join(output_dir, "train_stage1_preprocessed.csv"))
        layout_features = pd.read_csv(train_layout_path)
        final_train = pd.concat([merged_train, layout_features], axis=1)
        final_path = os.path.join(output_dir, "train_merged.csv")
        final_train.to_csv(final_path, index=False, encoding='utf_8_sig')
        print(f"âœ… åˆå¹¶åçš„è®­ç»ƒé›†å·²ä¿å­˜: {final_path}")
        return final_train

    final_train_df = update_merged_file()
    
    # éªŒè¯
    print("\n===== å¤„ç†ç»“æœæ ·ä¾‹ =====")
    display(final_train_df[['æˆ¿å±‹æˆ·å‹', 'æˆ·å‹_å§å®¤æ•°', 'æˆ·å‹_å«ç”Ÿé—´æ•°']].head(3))
    
except Exception as e:
    print(f"âŒ å¤„ç†è¿‡ç¨‹ä¸­å‡ºé”™: {e}")
    print("è¯·æ£€æŸ¥ä»¥ä¸‹è·¯å¾„æ˜¯å¦å­˜åœ¨:")
    print(f"- è¾“å…¥æ–‡ä»¶: {os.path.join(output_dir, 'train_stage1_preprocessed.csv')}")
    print(f"- è¾“å‡ºç›®å½•: {columns_dir}")
```

```python
def preprocess_floor(df):
    """
    ä¿®æ­£ç‰ˆçš„æ‰€åœ¨æ¥¼å±‚åˆ—é¢„å¤„ç†å‡½æ•°
    """
    df = df.copy()
    
    # æ›´å…¨é¢çš„å­—ç¬¦æ¸…ç†
    df['æ‰€åœ¨æ¥¼å±‚'] = df['æ‰€åœ¨æ¥¼å±‚'].str.replace(r'[â‘#\-Â·]', '', regex=True).str.strip()
    
    # æ›´çµæ´»çš„æ­£åˆ™è¡¨è¾¾å¼ï¼ˆåŒ¹é…ä¸­æ–‡/è‹±æ–‡æ‹¬å·ï¼‰
    pattern = r'([^\sï¼ˆ(]+)\s*[ï¼ˆ(]å…±\s*(\d+)\s*å±‚[)ï¼‰]'
    extracted = df['æ‰€åœ¨æ¥¼å±‚'].str.extract(pattern)
    
    # æå–æ¥¼å±‚ç±»å‹ï¼ˆæ›´å®½æ¾çš„å¤„ç†ï¼‰
    df['æ¥¼å±‚ç±»å‹'] = extracted[0].str.strip()
    df['æ€»æ¥¼å±‚æ•°'] = extracted[1].fillna('0').astype(int)
    
    # æ ‡å‡†åŒ–æ¥¼å±‚ç±»å‹è¡¨è¿°
    floor_type_mapping = {
        'ä½æ¥¼å±‚': 'ä½æ¥¼å±‚', 'ä½å±‚': 'ä½æ¥¼å±‚',
        'ä¸­æ¥¼å±‚': 'ä¸­æ¥¼å±‚', 'ä¸­å±‚': 'ä¸­æ¥¼å±‚',
        'é«˜æ¥¼å±‚': 'é«˜æ¥¼å±‚', 'é«˜å±‚': 'é«˜æ¥¼å±‚',
        'é¡¶æ¥¼': 'é¡¶å±‚', 'é¡¶æ¥¼å±‚': 'é¡¶å±‚', 'é¡¶å±‚': 'é¡¶å±‚',
        'åº•å±‚': 'åº•å±‚', 'åº•æ¥¼': 'åº•å±‚',
        'åœ°ä¸‹å®¤': 'åœ°ä¸‹å®¤', 'åœ°ä¸‹å±‚': 'åœ°ä¸‹å®¤'
    }
    df['æ¥¼å±‚ç±»å‹'] = df['æ¥¼å±‚ç±»å‹'].map(floor_type_mapping).fillna('ä¸­æ¥¼å±‚')  # é»˜è®¤å€¼
    
    # é‡æ–°è®¡ç®—ç›¸å¯¹ä½ç½®ï¼ˆæ·»åŠ æ›´å¤šä¸šåŠ¡è§„åˆ™ï¼‰
    df['ç›¸å¯¹ä½ç½®'] = df.apply(lambda x: calculate_relative_position(x['æ¥¼å±‚ç±»å‹'], x['æ€»æ¥¼å±‚æ•°']), axis=1)
    
    # åˆ›å»ºè¡ç”Ÿç‰¹å¾
    df['æ˜¯å¦åœ°ä¸‹å®¤'] = (df['æ¥¼å±‚ç±»å‹'] == 'åœ°ä¸‹å®¤').astype(int)
    df['æ˜¯å¦é¡¶å±‚'] = (df['æ¥¼å±‚ç±»å‹'] == 'é¡¶å±‚').astype(int)
    df['æ˜¯å¦åº•å±‚'] = (df['æ¥¼å±‚ç±»å‹'] == 'åº•å±‚').astype(int)
    
    # æ›´åˆç†çš„å»ºç­‘é«˜åº¦åˆ†ç®±
    bins = [-1, 3, 7, 18, 33, 100]  # è°ƒæ•´åˆ†ç®±è¾¹ç•Œ
    labels = ['ä½å±‚', 'å¤šå±‚', 'å°é«˜å±‚', 'é«˜å±‚', 'è¶…é«˜å±‚']
    df['å»ºç­‘é«˜åº¦ç±»å‹'] = pd.cut(df['æ€»æ¥¼å±‚æ•°'], bins=bins, labels=labels)
    
    return df

def calculate_relative_position(floor_type, total_floor):
    """æ›´ç²¾ç¡®çš„ç›¸å¯¹ä½ç½®è®¡ç®—"""
    position_map = {
        'åœ°ä¸‹å®¤': 0.0,
        'åº•å±‚': 0.15 if total_floor > 6 else 0.1,
        'ä½æ¥¼å±‚': 0.25 + min(total_floor/100, 0.1),
        'ä¸­æ¥¼å±‚': 0.5 + min(total_floor/200, 0.15),
        'é«˜æ¥¼å±‚': 0.75 - max((total_floor-30)/100, 0),
        'é¡¶å±‚': 1.0 - (0.05 if total_floor > 30 else 0)
    }
    return position_map.get(floor_type, 0.5)

    # é‡æ–°åŠ è½½åŸå§‹æ•°æ®ï¼ˆé¿å…é“¾å¼å¤„ç†ï¼‰
train_df = pd.read_csv("/home/mw/input/quant4533/ruc_Class25Q1_train.csv")

# åº”ç”¨ä¿®æ­£åçš„é¢„å¤„ç†
train_floor_processed = preprocess_floor(train_df)

# éªŒè¯å…³é”®å­—æ®µ
print("===== ä¿®æ­£åéªŒè¯ =====")
print("1. æˆåŠŸæå–è®°å½•å æ¯”:", 
      f"{len(train_floor_processed[train_floor_processed['æ€»æ¥¼å±‚æ•°']>0])/len(train_floor_processed):.1%}")

print("\n2. æ¥¼å±‚ç±»å‹åˆ†å¸ƒ:")
print(train_floor_processed['æ¥¼å±‚ç±»å‹'].value_counts())

print("\n3. æ ·ä¾‹æ•°æ®:")
sample = train_floor_processed[['æ‰€åœ¨æ¥¼å±‚', 'æ¥¼å±‚ç±»å‹', 'æ€»æ¥¼å±‚æ•°', 'ç›¸å¯¹ä½ç½®',]].head(5)
display(sample)
```

```python
def preprocess_area(df):
    """
    å»ºç­‘é¢ç§¯å’Œå¥—å†…é¢ç§¯é¢„å¤„ç†å‡½æ•°
    å¤„ç†é€»è¾‘ï¼š
    1. æ¸…é™¤é¢ç§¯å•ä½ç¬¦å·
    2. å¤„ç†ç¼ºå¤±å€¼ï¼ˆåŸºäºå¾—æˆ¿ç‡ä¼°ç®—ï¼‰
    3. åˆ›å»ºè¡ç”Ÿç‰¹å¾
    4. å¼‚å¸¸å€¼å¤„ç†
    """
    df = df.copy()
    
    # æ¸…é™¤é¢ç§¯å•ä½ç¬¦å·ï¼ˆã¡ï¼‰å¹¶è½¬æ¢ä¸ºæ•°å€¼
    df['å»ºç­‘é¢ç§¯'] = df['å»ºç­‘é¢ç§¯'].str.replace('ã¡', '').astype(float)
    df['å¥—å†…é¢ç§¯'] = df['å¥—å†…é¢ç§¯'].str.replace('ã¡', '').astype(float)
    
    # è®¡ç®—å¾—æˆ¿ç‡ï¼ˆå¥—å†…/å»ºç­‘ï¼‰
    df['å¾—æˆ¿ç‡'] = df['å¥—å†…é¢ç§¯'] / df['å»ºç­‘é¢ç§¯']
    
    # å¤„ç†ç¼ºå¤±çš„å¥—å†…é¢ç§¯ï¼ˆç”¨å»ºç­‘é¢ç§¯*ä¸­ä½æ•°å¾—æˆ¿ç‡ï¼‰
    median_ratio = df['å¾—æˆ¿ç‡'].median()
    df['å¥—å†…é¢ç§¯'] = df['å¥—å†…é¢ç§¯'].fillna(df['å»ºç­‘é¢ç§¯'] * median_ratio)
    
    # å¤„ç†å¼‚å¸¸å€¼
    df.loc[df['å»ºç­‘é¢ç§¯'] < 15, 'å»ºç­‘é¢ç§¯'] = 15  # æœ€å°åˆç†é¢ç§¯
    df.loc[df['å¥—å†…é¢ç§¯'] > df['å»ºç­‘é¢ç§¯'], 'å¥—å†…é¢ç§¯'] = df['å»ºç­‘é¢ç§¯'] * 0.9  # å¥—å†…ä¸åº”å¤§äºå»ºç­‘é¢ç§¯
    
    # é‡æ–°è®¡ç®—å¾—æˆ¿ç‡
    df['å¾—æˆ¿ç‡'] = df['å¥—å†…é¢ç§¯'] / df['å»ºç­‘é¢ç§¯']
    
    # åˆ›å»ºè¡ç”Ÿç‰¹å¾
    df['å…¬æ‘Šé¢ç§¯'] = df['å»ºç­‘é¢ç§¯'] - df['å¥—å†…é¢ç§¯']
    df['é¢ç§¯ç­‰çº§'] = pd.cut(df['å»ºç­‘é¢ç§¯'],
                         bins=[0, 50, 90, 144, 200, 300, 1000],
                         labels=['æå°', 'å°', 'ä¸­', 'å¤§', 'è¶…å¤§', 'è±ªå®…'])
    df['æ˜¯å¦è±ªå®…'] = (df['å»ºç­‘é¢ç§¯'] > 144).astype(int)
    df['æ˜¯å¦å°æˆ·å‹'] = (df['å»ºç­‘é¢ç§¯'] < 50).astype(int)
    
    return df

# ==================== è®­ç»ƒé›†å¤„ç† ====================
# åŠ è½½ä¹‹å‰åˆå¹¶çš„æ•°æ®
train_merged_path = f"{output_dir}/train_merged_v2.csv"
train_df = pd.read_csv(train_merged_path)

# åº”ç”¨é¢ç§¯é¢„å¤„ç†
train_area_processed = preprocess_area(train_df)

# éªŒè¯å¤„ç†ç»“æœ
print("===== é¢ç§¯å¤„ç†éªŒè¯ =====")
print("1. ç¼ºå¤±å€¼å¤„ç†ç»“æœ:")
print(" - å¥—å†…é¢ç§¯ç¼ºå¤±å€¼:", train_area_processed['å¥—å†…é¢ç§¯'].isnull().sum())
print(" - å¾—æˆ¿ç‡ä¸­ä½æ•°:", round(train_area_processed['å¾—æˆ¿ç‡'].median(), 3))

print("\n2. é¢ç§¯ç»Ÿè®¡æè¿°:")
print(train_area_processed[['å»ºç­‘é¢ç§¯', 'å¥—å†…é¢ç§¯', 'å¾—æˆ¿ç‡']].describe())

print("\n3. é¢ç§¯ç­‰çº§åˆ†å¸ƒ:")
print(train_area_processed['é¢ç§¯ç­‰çº§'].value_counts())

# ä¿å­˜é¢ç§¯ç‰¹å¾ï¼ˆä»…æ–°ç”Ÿæˆçš„åˆ—ï¼‰
area_feature_cols = ['å»ºç­‘é¢ç§¯', 'å¥—å†…é¢ç§¯', 'å¾—æˆ¿ç‡', 'å…¬æ‘Šé¢ç§¯', 
                    'é¢ç§¯ç­‰çº§', 'æ˜¯å¦è±ªå®…', 'æ˜¯å¦å°æˆ·å‹']
train_area_features = train_area_processed[area_feature_cols]

# æ·»åŠ å‰ç¼€ä¾¿äºåç»­åˆå¹¶
train_area_features = train_area_features.add_prefix('é¢ç§¯_')

# ä¿å­˜åˆ°columnsç›®å½•
train_area_path = f"{output_dir}/columns/train_é¢ç§¯.csv"
train_area_features.to_csv(train_area_path, index=False, encoding='utf_8_sig')
print(f"\nğŸ’¾ é¢ç§¯ç‰¹å¾å·²ä¿å­˜åˆ°: {train_area_path}")

# æ›´æ–°åˆå¹¶æ–‡ä»¶
def update_train_merged():
    merged_train = pd.read_csv(train_merged_path)
    area_features = pd.read_csv(train_area_path)
    final_train = pd.concat([merged_train, area_features], axis=1)
    
    final_path = f"{output_dir}/train_merged_v3.csv"  # æ›´æ–°ç‰ˆæœ¬å·
    final_train.to_csv(final_path, index=False, encoding='utf_8_sig')
    print(f"ğŸ”„ åˆå¹¶åçš„è®­ç»ƒé›†å·²æ›´æ–°: {final_path}")
    return final_train

final_train_df = update_train_merged()

# æ˜¾ç¤ºå¤„ç†æ ·ä¾‹
print("\n===== å¤„ç†ç»“æœæ ·ä¾‹ =====")
sample_cols = ['å»ºç­‘é¢ç§¯', 'å¥—å†…é¢ç§¯', 'é¢ç§¯_å¾—æˆ¿ç‡', 'é¢ç§¯_é¢ç§¯ç­‰çº§']
display(final_train_df[sample_cols].head(5))
```

```python
# åŠ è½½æµ‹è¯•é›†åˆå¹¶æ•°æ®
test_merged_path = f"{output_dir}/test_merged_v2.csv"
test_df = pd.read_csv(test_merged_path)

# åº”ç”¨ç›¸åŒçš„é¢„å¤„ç†å‡½æ•°ï¼ˆä½¿ç”¨è®­ç»ƒé›†è®¡ç®—çš„å¾—æˆ¿ç‡ä¸­ä½æ•°ï¼‰
def preprocess_test_area(df, train_median_ratio):
    """
    æµ‹è¯•é›†é¢ç§¯é¢„å¤„ç†ï¼ˆä¿æŒä¸è®­ç»ƒé›†ä¸€è‡´çš„å¤„ç†é€»è¾‘ï¼‰
    å‚æ•°:
        df: æµ‹è¯•é›†DataFrame
        train_median_ratio: è®­ç»ƒé›†è®¡ç®—çš„å¾—æˆ¿ç‡ä¸­ä½æ•°
    """
    df = df.copy()
    
    # ç»Ÿä¸€å•ä½å¤„ç†
    df['å»ºç­‘é¢ç§¯'] = df['å»ºç­‘é¢ç§¯'].str.replace('ã¡', '').astype(float)
    df['å¥—å†…é¢ç§¯'] = df['å¥—å†…é¢ç§¯'].str.replace('ã¡', '').astype(float)
    
    # ä½¿ç”¨è®­ç»ƒé›†çš„å¾—æˆ¿ç‡ä¸­ä½æ•°å¡«å……ç¼ºå¤±å€¼ï¼ˆå…³é”®ä¸€è‡´æ€§ä¿éšœï¼‰
    df['å¥—å†…é¢ç§¯'] = df['å¥—å†…é¢ç§¯'].fillna(df['å»ºç­‘é¢ç§¯'] * train_median_ratio)
    
    # åº”ç”¨ç›¸åŒçš„å¼‚å¸¸å€¼å¤„ç†é˜ˆå€¼
    df.loc[df['å»ºç­‘é¢ç§¯'] < 15, 'å»ºç­‘é¢ç§¯'] = 15
    df.loc[df['å¥—å†…é¢ç§¯'] > df['å»ºç­‘é¢ç§¯'], 'å¥—å†…é¢ç§¯'] = df['å»ºç­‘é¢ç§¯'] * 0.9
    
    # è®¡ç®—å¾—æˆ¿ç‡
    df['å¾—æˆ¿ç‡'] = df['å¥—å†…é¢ç§¯'] / df['å»ºç­‘é¢ç§¯']
    
    # ç›¸åŒçš„è¡ç”Ÿç‰¹å¾
    df['å…¬æ‘Šé¢ç§¯'] = df['å»ºç­‘é¢ç§¯'] - df['å¥—å†…é¢ç§¯']
    df['é¢ç§¯ç­‰çº§'] = pd.cut(df['å»ºç­‘é¢ç§¯'],
                         bins=[0, 50, 90, 144, 200, 300, 1000],
                         labels=['æå°', 'å°', 'ä¸­', 'å¤§', 'è¶…å¤§', 'è±ªå®…'])
    df['æ˜¯å¦è±ªå®…'] = (df['å»ºç­‘é¢ç§¯'] > 144).astype(int)
    df['æ˜¯å¦å°æˆ·å‹'] = (df['å»ºç­‘é¢ç§¯'] < 50).astype(int)
    
    return df

# è·å–è®­ç»ƒé›†çš„å¾—æˆ¿ç‡ä¸­ä½æ•°ï¼ˆå…³é”®æ­¥éª¤ï¼ï¼‰
train_median_ratio = pd.read_csv(f"{output_dir}/columns/train_é¢ç§¯.csv")['é¢ç§¯_å¾—æˆ¿ç‡'].median()

# åº”ç”¨é¢„å¤„ç†
test_area_processed = preprocess_test_area(test_df, train_median_ratio)

# éªŒè¯å¤„ç†ä¸€è‡´æ€§
print("===== æµ‹è¯•é›†é¢ç§¯å¤„ç†éªŒè¯ =====")
print("1. ä½¿ç”¨çš„è®­ç»ƒé›†å¾—æˆ¿ç‡ä¸­ä½æ•°:", round(train_median_ratio, 3))
print("2. æµ‹è¯•é›†å¾—æˆ¿ç‡ä¸­ä½æ•°:", round(test_area_processed['å¾—æˆ¿ç‡'].median(), 3))

print("\n3. é¢ç§¯ç»Ÿè®¡æè¿°:")
print(test_area_processed[['å»ºç­‘é¢ç§¯', 'å¥—å†…é¢ç§¯', 'å¾—æˆ¿ç‡']].describe())

print("\n4. é¢ç§¯ç­‰çº§åˆ†å¸ƒ:")
print(test_area_processed['é¢ç§¯ç­‰çº§'].value_counts())

# ä¿å­˜é¢ç§¯ç‰¹å¾ï¼ˆä»…æ–°ç”Ÿæˆçš„åˆ—ï¼‰
test_area_features = test_area_processed[[
    'å»ºç­‘é¢ç§¯', 'å¥—å†…é¢ç§¯', 'å¾—æˆ¿ç‡', 'å…¬æ‘Šé¢ç§¯', 
    'é¢ç§¯ç­‰çº§', 'æ˜¯å¦è±ªå®…', 'æ˜¯å¦å°æˆ·å‹'
]].add_prefix('é¢ç§¯_')

# ä¿å­˜åˆ°columnsç›®å½•
test_area_path = f"{output_dir}/columns/test_é¢ç§¯.csv"
test_area_features.to_csv(test_area_path, index=False, encoding='utf_8_sig')
print(f"\nğŸ’¾ æµ‹è¯•é›†é¢ç§¯ç‰¹å¾å·²ä¿å­˜åˆ°: {test_area_path}")

# æ›´æ–°æµ‹è¯•é›†åˆå¹¶æ–‡ä»¶
def update_test_merged():
    merged_test = pd.read_csv(test_merged_path)
    area_features = pd.read_csv(test_area_path)
    final_test = pd.concat([merged_test, area_features], axis=1)
    
    final_path = f"{output_dir}/test_merged_v3.csv"
    final_test.to_csv(final_path, index=False, encoding='utf_8_sig')
    print(f"ğŸ”„ åˆå¹¶åçš„æµ‹è¯•é›†å·²æ›´æ–°: {final_path}")
    return final_test

final_test_df = update_test_merged()

# æ˜¾ç¤ºå¤„ç†æ ·ä¾‹
print("\n===== æµ‹è¯•é›†å¤„ç†ç»“æœæ ·ä¾‹ =====")
sample_cols = ['å»ºç­‘é¢ç§¯', 'å¥—å†…é¢ç§¯', 'é¢ç§¯_å¾—æˆ¿ç‡', 'é¢ç§¯_æ˜¯å¦è±ªå®…']
display(final_test_df[sample_cols].head(5))
```

```python
def preprocess_orientation(df):
    """
    æˆ¿å±‹æœå‘åˆ—é¢„å¤„ç†å‡½æ•°
    å¤„ç†é€»è¾‘ï¼š
    1. æ ‡å‡†åŒ–æ–¹å‘è¡¨è¿°ï¼ˆå»é™¤ç‰¹æ®Šç¬¦å·ã€ç»Ÿä¸€æ ¼å¼ï¼‰
    2. æå–ä¸»æœå‘å’Œç»„åˆç‰¹å¾
    3. åˆ›å»ºè¡ç”Ÿç‰¹å¾
    """
    df = df.copy()
    
    # ç»Ÿä¸€æ ¼å¼å¤„ç†ï¼ˆå»é™¤ç¬¦å·ã€ç©ºæ ¼æ ‡å‡†åŒ–ï¼‰
    df['æˆ¿å±‹æœå‘'] = df['æˆ¿å±‹æœå‘'].str.replace(r'[â–¡\s]', '', regex=True).str.strip()
    
    # æå–åŸºç¡€æ–¹å‘ï¼ˆ8ä¸ªä¸»è¦æ–¹ä½ï¼‰
    base_directions = ['ä¸œ', 'å—', 'è¥¿', 'åŒ—', 'ä¸œå—', 'ä¸œåŒ—', 'è¥¿å—', 'è¥¿åŒ—']
    
    # åˆ›å»ºå„æ–¹å‘çš„äºŒå…ƒç‰¹å¾
    for direction in base_directions:
        df[f'æœå‘_{direction}'] = df['æˆ¿å±‹æœå‘'].apply(lambda x: 1 if direction in x else 0)
    
    # åˆ¤æ–­æ˜¯å¦å¤šæœå‘ç»„åˆ
    df['æœå‘_ç»„åˆæ•°'] = df['æˆ¿å±‹æœå‘'].apply(lambda x: len([d for d in base_directions if d in x]))
    
    # æå–ä¸»æœå‘ï¼ˆæŒ‰ä»·å€¼æ’åºï¼šå— > ä¸œå— > ä¸œ > è¥¿å— > åŒ— > è¥¿ > ä¸œåŒ— > è¥¿åŒ—ï¼‰
    priority_order = ['å—', 'ä¸œå—', 'ä¸œ', 'è¥¿å—', 'åŒ—', 'è¥¿', 'ä¸œåŒ—', 'è¥¿åŒ—']
    def get_main_direction(x):
        for direction in priority_order:
            if direction in x:
                return direction
        return 'å…¶ä»–'
    df['æœå‘_ä¸»æ–¹å‘'] = df['æˆ¿å±‹æœå‘'].apply(get_main_direction)
    
    # åˆ›å»ºä¼˜è´¨æœå‘æ ‡å¿—
    df['æœå‘_ä¼˜è´¨'] = df['æœå‘_ä¸»æ–¹å‘'].isin(['å—', 'ä¸œå—', 'ä¸œ']).astype(int)
    
    # ç‰¹æ®Šç»„åˆç‰¹å¾
    df['æœå‘_å—åŒ—é€šé€'] = (df['æœå‘_å—'] & df['æœå‘_åŒ—']).astype(int)
    df['æœå‘_ä¸œè¥¿é€šé€'] = (df['æœå‘_ä¸œ'] & df['æœå‘_è¥¿']).astype(int)
    
    return df

# ==================== è®­ç»ƒé›†å¤„ç† ====================
# åŠ è½½ä¹‹å‰åˆå¹¶çš„æ•°æ®
train_merged_path = f"{output_dir}/train_merged_v3.csv"
train_df = pd.read_csv(train_merged_path)

# åº”ç”¨æœå‘é¢„å¤„ç†
train_orientation_processed = preprocess_orientation(train_df)

# éªŒè¯å¤„ç†ç»“æœ
print("===== æœå‘å¤„ç†éªŒè¯ =====")
print("1. ä¸»æ–¹å‘åˆ†å¸ƒ:")
print(train_orientation_processed['æœå‘_ä¸»æ–¹å‘'].value_counts())

print("\n2. ç»„åˆæ•°åˆ†å¸ƒ:")
print(train_orientation_processed['æœå‘_ç»„åˆæ•°'].value_counts().sort_index())

print("\n3. ä¼˜è´¨æœå‘å æ¯”:", 
      f"{train_orientation_processed['æœå‘_ä¼˜è´¨'].mean():.1%}")

# ä¿å­˜æœå‘ç‰¹å¾ï¼ˆä»…æ–°ç”Ÿæˆçš„åˆ—ï¼‰
orientation_cols = [col for col in train_orientation_processed.columns 
                   if col.startswith('æœå‘_')]
train_orientation_features = train_orientation_processed[orientation_cols]

# ä¿å­˜åˆ°columnsç›®å½•
train_orientation_path = f"{output_dir}/columns/train_æˆ¿å±‹æœå‘.csv"
train_orientation_features.to_csv(train_orientation_path, index=False, encoding='utf_8_sig')
print(f"\nğŸ’¾ æœå‘ç‰¹å¾å·²ä¿å­˜åˆ°: {train_orientation_path}")

# æ›´æ–°åˆå¹¶æ–‡ä»¶
def update_train_merged():
    merged_train = pd.read_csv(train_merged_path)
    orientation_features = pd.read_csv(train_orientation_path)
    final_train = pd.concat([merged_train, orientation_features], axis=1)
    
    final_path = f"{output_dir}/train_merged_v4.csv"
    final_train.to_csv(final_path, index=False, encoding='utf_8_sig')
    print(f"ğŸ”„ åˆå¹¶åçš„è®­ç»ƒé›†å·²æ›´æ–°: {final_path}")
    return final_train

final_train_df = update_train_merged()

# æ˜¾ç¤ºå¤„ç†æ ·ä¾‹
print("\n===== å¤„ç†ç»“æœæ ·ä¾‹ =====")
sample_cols = ['æˆ¿å±‹æœå‘', 'æœå‘_ä¸»æ–¹å‘', 'æœå‘_ç»„åˆæ•°', 'æœå‘_ä¼˜è´¨', 'æœå‘_å—åŒ—é€šé€']
display(final_train_df[sample_cols].head(8))
```

```python
#trainçš„å»ºç­‘ç»“æ„
def preprocess_structure(df):
    """
    å»ºç­‘ç»“æ„åˆ—é¢„å¤„ç†å‡½æ•°
    å¤„ç†é€»è¾‘ï¼š
    1. å¡«è¡¥ç©ºç™½å€¼ä¸ºæœ€å¸¸è§ç±»å‹ï¼ˆç –æ··ç»“æ„ï¼‰
    2. æœ‰åºç¼–ç ï¼ˆæŒ‰ç»“æ„è´¨é‡ä»é«˜åˆ°ä½ï¼‰
    3. åˆ›å»ºè¡ç”Ÿç‰¹å¾
    """
    df = df.copy()
    
    # å¡«è¡¥ç©ºç™½å€¼
    df['å»ºç­‘ç»“æ„'] = df['å»ºç­‘ç»“æ„'].fillna('ç –æ··ç»“æ„')
    
    # ç»Ÿä¸€è¡¨è¿°
    df['å»ºç­‘ç»“æ„'] = df['å»ºç­‘ç»“æ„'].str.replace('é’¢æ··ç»“æ„', 'é’¢ç­‹æ··å‡åœŸç»“æ„')
    
    # æœ‰åºç¼–ç å­—å…¸ï¼ˆæ•°å€¼è¶Šå¤§è¡¨ç¤ºç»“æ„è´¨é‡è¶Šå¥½ï¼‰
    structure_order = {
        'é’¢ç»“æ„': 4,           # æœ€é«˜çº§
        'é’¢ç­‹æ··å‡åœŸç»“æ„': 3,   # ç°ä»£ä¸»æµ
        'æ¡†æ¶ç»“æ„': 2,         # å¸¸è§å°é«˜å±‚
        'æ··åˆç»“æ„': 1,         # è¿‡æ¸¡ç±»å‹
        'ç –æ··ç»“æ„': 0,         # è€æ—§ä½å®…
        'ç –æœ¨ç»“æ„': -1,        # ä»·å€¼æœ€ä½
        'æœªçŸ¥ç»“æ„': 0          # é»˜è®¤æŒ‰ç –æ··å¤„ç†
    }
    
    # åº”ç”¨ç¼–ç 
    df['ç»“æ„_ç¼–ç '] = df['å»ºç­‘ç»“æ„'].map(structure_order)
    
    # åˆ›å»ºè¡ç”Ÿç‰¹å¾
    df['ç»“æ„_ç°ä»£å»ºç­‘'] = (df['ç»“æ„_ç¼–ç '] >= 2).astype(int)  # é’¢/é’¢ç­‹æ··å‡åœŸ/æ¡†æ¶
    df['ç»“æ„_éœ€è­¦æƒ•'] = (df['ç»“æ„_ç¼–ç '] <= -1).astype(int)   # ç –æœ¨ç»“æ„
    
    # ä¿å­˜ç¼–ç æ˜ å°„
    pd.DataFrame({
        'åŸå§‹å€¼': list(structure_order.keys()),
        'ç¼–ç å€¼': list(structure_order.values())
    }).to_csv(f"{output_dir}/structure_encoding_map.csv", index=False)
    
    return df

# ==================== è®­ç»ƒé›†å¤„ç† ====================
# åŠ è½½ä¹‹å‰åˆå¹¶çš„æ•°æ®
train_merged_path = f"{output_dir}/train_merged_v4.csv"
train_df = pd.read_csv(train_merged_path)

# åº”ç”¨ç»“æ„é¢„å¤„ç†
train_structure_processed = preprocess_structure(train_df)

# éªŒè¯å¤„ç†ç»“æœ
print("===== å»ºç­‘ç»“æ„å¤„ç†éªŒè¯ =====")
print("1. ç»“æ„ç±»å‹åˆ†å¸ƒ:")
print(train_structure_processed['å»ºç­‘ç»“æ„'].value_counts())

print("\n2. ç¼–ç å€¼åˆ†å¸ƒ:")
print(train_structure_processed['ç»“æ„_ç¼–ç '].value_counts().sort_index())

print("\n3. è¡ç”Ÿç‰¹å¾ç»Ÿè®¡:")
print("ç°ä»£å»ºç­‘å æ¯”:", f"{train_structure_processed['ç»“æ„_ç°ä»£å»ºç­‘'].mean():.1%}")
print("éœ€è­¦æƒ•ç»“æ„å æ¯”:", f"{train_structure_processed['ç»“æ„_éœ€è­¦æƒ•'].mean():.1%}")

# ä¿å­˜ç»“æ„ç‰¹å¾ï¼ˆä»…æ–°ç”Ÿæˆçš„åˆ—ï¼‰
structure_cols = ['ç»“æ„_ç¼–ç ', 'ç»“æ„_ç°ä»£å»ºç­‘', 'ç»“æ„_éœ€è­¦æƒ•']
train_structure_features = train_structure_processed[structure_cols]

# ä¿å­˜åˆ°columnsç›®å½•
train_structure_path = f"{output_dir}/columns/train_å»ºç­‘ç»“æ„.csv"
train_structure_features.to_csv(train_structure_path, index=False, encoding='utf_8_sig')
print(f"\nğŸ’¾ ç»“æ„ç‰¹å¾å·²ä¿å­˜åˆ°: {train_structure_path}")

# æ›´æ–°åˆå¹¶æ–‡ä»¶
def update_train_merged():
    merged_train = pd.read_csv(train_merged_path)
    structure_features = pd.read_csv(train_structure_path)
    final_train = pd.concat([merged_train, structure_features], axis=1)
    
    final_path = f"{output_dir}/train_merged_v5.csv"
    final_train.to_csv(final_path, index=False, encoding='utf_8_sig')
    print(f"ğŸ”„ åˆå¹¶åçš„è®­ç»ƒé›†å·²æ›´æ–°: {final_path}")
    return final_train

final_train_df = update_train_merged()

# æ˜¾ç¤ºå¤„ç†æ ·ä¾‹
print("\n===== å¤„ç†ç»“æœæ ·ä¾‹ =====")
sample_cols = ['å»ºç­‘ç»“æ„', 'ç»“æ„_ç¼–ç ', 'ç»“æ„_ç°ä»£å»ºç­‘']
display(final_train_df[sample_cols].head(5))
```

```python
# åŠ è½½æµ‹è¯•é›†åˆå¹¶æ•°æ®
test_merged_path = f"{output_dir}/test_merged_v4.csv"
test_df = pd.read_csv(test_merged_path)

# åº”ç”¨ç›¸åŒçš„é¢„å¤„ç†ï¼ˆä½¿ç”¨è®­ç»ƒé›†ç”Ÿæˆçš„ç¼–ç æ˜ å°„ï¼‰
def preprocess_test_structure(df):
    """
    æµ‹è¯•é›†å»ºç­‘ç»“æ„é¢„å¤„ç†ï¼ˆä¿æŒä¸è®­ç»ƒé›†ä¸€è‡´ï¼‰
    å‚æ•°:
        df: æµ‹è¯•é›†DataFrame
    è¿”å›:
        å¤„ç†åçš„DataFrameï¼ˆä»…åŒ…å«ç»“æ„ç›¸å…³ç‰¹å¾ï¼‰
    """
    # åŠ è½½è®­ç»ƒé›†åˆ›å»ºçš„ç¼–ç æ˜ å°„
    structure_map = pd.read_csv(f"{output_dir}/structure_encoding_map.csv")
    structure_order = dict(zip(structure_map['åŸå§‹å€¼'], structure_map['ç¼–ç å€¼']))
    
    df = df.copy()
    df['å»ºç­‘ç»“æ„'] = df['å»ºç­‘ç»“æ„'].fillna('ç –æ··ç»“æ„').str.replace('é’¢æ··ç»“æ„', 'é’¢ç­‹æ··å‡åœŸç»“æ„')
    
    # åº”ç”¨ç¼–ç ï¼ˆæœªçŸ¥å€¼è®¾ä¸ºç –æ··ç»“æ„çš„ç¼–ç 0ï¼‰
    df['ç»“æ„_ç¼–ç '] = df['å»ºç­‘ç»“æ„'].map(structure_order).fillna(0)
    
    # ç›¸åŒçš„è¡ç”Ÿç‰¹å¾
    df['ç»“æ„_ç°ä»£å»ºç­‘'] = (df['ç»“æ„_ç¼–ç '] >= 2).astype(int)
    df['ç»“æ„_éœ€è­¦æƒ•'] = (df['ç»“æ„_ç¼–ç '] <= -1).astype(int)
    
    return df[['ç»“æ„_ç¼–ç ', 'ç»“æ„_ç°ä»£å»ºç­‘', 'ç»“æ„_éœ€è­¦æƒ•']]

# åº”ç”¨é¢„å¤„ç†
test_structure_features = preprocess_test_structure(test_df)

# éªŒè¯å¤„ç†ä¸€è‡´æ€§
print("===== æµ‹è¯•é›†ç»“æ„å¤„ç†éªŒè¯ =====")
print("1. ç¼–ç å€¼åˆ†å¸ƒ:")
print(test_structure_features['ç»“æ„_ç¼–ç '].value_counts().sort_index())

print("\n2. è¡ç”Ÿç‰¹å¾ç»Ÿè®¡:")
print("ç°ä»£å»ºç­‘å æ¯”:", f"{test_structure_features['ç»“æ„_ç°ä»£å»ºç­‘'].mean():.1%}")
print("éœ€è­¦æƒ•ç»“æ„å æ¯”:", f"{test_structure_features['ç»“æ„_éœ€è­¦æƒ•'].mean():.1%}")

# ä¿å­˜åˆ°columnsç›®å½•
test_structure_path = f"{output_dir}/columns/test_å»ºç­‘ç»“æ„.csv"
test_structure_features.to_csv(test_structure_path, index=False, encoding='utf_8_sig')
print(f"\nğŸ’¾ æµ‹è¯•é›†ç»“æ„ç‰¹å¾å·²ä¿å­˜åˆ°: {test_structure_path}")

# æ›´æ–°æµ‹è¯•é›†åˆå¹¶æ–‡ä»¶
def update_test_merged():
    merged_test = pd.read_csv(test_merged_path)
    structure_features = pd.read_csv(test_structure_path)
    final_test = pd.concat([merged_test, structure_features], axis=1)
    
    final_path = f"{output_dir}/test_merged_v5.csv"
    final_test.to_csv(final_path, index=False, encoding='utf_8_sig')
    print(f"ğŸ”„ åˆå¹¶åçš„æµ‹è¯•é›†å·²æ›´æ–°: {final_path}")
    return final_test

final_test_df = update_test_merged()

# æ˜¾ç¤ºå¤„ç†æ ·ä¾‹
print("\n===== æµ‹è¯•é›†å¤„ç†ç»“æœæ ·ä¾‹ =====")
sample_cols = ['å»ºç­‘ç»“æ„', 'ç»“æ„_ç¼–ç ', 'ç»“æ„_ç°ä»£å»ºç­‘']
display(final_test_df[sample_cols].head(5))
```

```python
def preprocess_renovation(df):
    """
    è£…ä¿®æƒ…å†µåˆ—é¢„å¤„ç†å‡½æ•°
    å¤„ç†é€»è¾‘ï¼š
    1. å¡«è¡¥ç©ºç™½å€¼ä¸ºæœ€å¸¸è§ç±»å‹ï¼ˆç®€è£…ï¼‰
    2. æœ‰åºç¼–ç ï¼ˆæŒ‰è£…ä¿®è´¨é‡ä»é«˜åˆ°ä½ï¼‰
    3. åˆ›å»ºè¡ç”Ÿç‰¹å¾
    """
    df = df.copy()
    
    # å¡«è¡¥ç©ºç™½å€¼
    df['è£…ä¿®æƒ…å†µ'] = df['è£…ä¿®æƒ…å†µ'].fillna('ç®€è£…')
    
    # æœ‰åºç¼–ç å­—å…¸ï¼ˆæ•°å€¼è¶Šå¤§è¡¨ç¤ºè£…ä¿®è¶Šå¥½ï¼‰
    renovation_order = {
        'ç²¾è£…': 3,    # æœ€é«˜æ ‡å‡†
        'ç®€è£…': 1,     # åŸºç¡€è£…ä¿®
        'æ¯›å¯': 0,     # æ— è£…ä¿®
        'å…¶ä»–': 1      # é»˜è®¤æŒ‰ç®€è£…å¤„ç†
    }
    
    # åº”ç”¨ç¼–ç 
    df['è£…ä¿®_ç¼–ç '] = df['è£…ä¿®æƒ…å†µ'].map(renovation_order)
    
    # åˆ›å»ºè¡ç”Ÿç‰¹å¾
    df['è£…ä¿®_ç²¾è£…ä¿®'] = (df['è£…ä¿®_ç¼–ç '] == 3).astype(int)
    df['è£…ä¿®_éœ€è£…ä¿®'] = (df['è£…ä¿®_ç¼–ç '] == 0).astype(int)
    
    # ä¿å­˜ç¼–ç æ˜ å°„
    pd.DataFrame({
        'åŸå§‹å€¼': list(renovation_order.keys()),
        'ç¼–ç å€¼': list(renovation_order.values())
    }).to_csv(f"{output_dir}/renovation_encoding_map.csv", index=False)
    
    return df

# ==================== è®­ç»ƒé›†å¤„ç† ====================
# åŠ è½½ä¹‹å‰åˆå¹¶çš„æ•°æ®
train_merged_path = f"{output_dir}/train_merged_v5.csv"
train_df = pd.read_csv(train_merged_path)

# åº”ç”¨è£…ä¿®é¢„å¤„ç†
train_renovation_processed = preprocess_renovation(train_df)

# éªŒè¯å¤„ç†ç»“æœ
print("===== è£…ä¿®æƒ…å†µå¤„ç†éªŒè¯ =====")
print("1. è£…ä¿®ç±»å‹åˆ†å¸ƒ:")
print(train_renovation_processed['è£…ä¿®æƒ…å†µ'].value_counts())

print("\n2. ç¼–ç å€¼åˆ†å¸ƒ:")
print(train_renovation_processed['è£…ä¿®_ç¼–ç '].value_counts().sort_index())

print("\n3. è¡ç”Ÿç‰¹å¾ç»Ÿè®¡:")
print("ç²¾è£…ä¿®å æ¯”:", f"{train_renovation_processed['è£…ä¿®_ç²¾è£…ä¿®'].mean():.1%}")
print("éœ€è£…ä¿®å æ¯”:", f"{train_renovation_processed['è£…ä¿®_éœ€è£…ä¿®'].mean():.1%}")

# ä¿å­˜è£…ä¿®ç‰¹å¾ï¼ˆä»…æ–°ç”Ÿæˆçš„åˆ—ï¼‰
renovation_cols = ['è£…ä¿®_ç¼–ç ', 'è£…ä¿®_ç²¾è£…ä¿®', 'è£…ä¿®_éœ€è£…ä¿®']
train_renovation_features = train_renovation_processed[renovation_cols]

# ä¿å­˜åˆ°columnsç›®å½•
train_renovation_path = f"{output_dir}/columns/train_è£…ä¿®æƒ…å†µ.csv"
train_renovation_features.to_csv(train_renovation_path, index=False, encoding='utf_8_sig')
print(f"\nğŸ’¾ è£…ä¿®ç‰¹å¾å·²ä¿å­˜åˆ°: {train_renovation_path}")

# æ›´æ–°åˆå¹¶æ–‡ä»¶
def update_train_merged():
    merged_train = pd.read_csv(train_merged_path)
    renovation_features = pd.read_csv(train_renovation_path)
    final_train = pd.concat([merged_train, renovation_features], axis=1)
    
    final_path = f"{output_dir}/train_merged_v6.csv"
    final_train.to_csv(final_path, index=False, encoding='utf_8_sig')
    print(f"ğŸ”„ åˆå¹¶åçš„è®­ç»ƒé›†å·²æ›´æ–°: {final_path}")
    return final_train

final_train_df = update_train_merged()

# æ˜¾ç¤ºå¤„ç†æ ·ä¾‹
print("\n===== å¤„ç†ç»“æœæ ·ä¾‹ =====")
sample_cols = ['è£…ä¿®æƒ…å†µ', 'è£…ä¿®_ç¼–ç ', 'è£…ä¿®_ç²¾è£…ä¿®']
display(final_train_df[sample_cols].head(5))
```

```python
# ä¿®æ­£åçš„"æ¢¯æˆ·æ¯”ä¾‹"åˆ—é¢„å¤„ç†å®Œæ•´ä»£ç 

import pandas as pd
import os

# ç¡®ä¿è¾“å‡ºç›®å½•å­˜åœ¨
output_dir = "/home/mw/project/output"
os.makedirs(output_dir, exist_ok=True)

def preprocess_elevator_ratio(df):
    """
    æ¢¯æˆ·æ¯”ä¾‹åˆ—é¢„å¤„ç†å‡½æ•°ï¼ˆä¿®æ­£ç‰ˆï¼‰
    å¤„ç†é€»è¾‘ï¼š
    1. æå–ç”µæ¢¯æ•°é‡å’Œæ¯æ¢¯æˆ·æ•°
    2. è®¡ç®—ç”µæ¢¯æœåŠ¡å¯†åº¦
    3. åˆ›å»ºè¡ç”Ÿç‰¹å¾ï¼ˆç¡®ä¿åˆ—åæ­£ç¡®ï¼‰
    """
    df = df.copy()
    
    # ç»Ÿä¸€æ ¼å¼å¤„ç†ï¼ˆå»é™¤ç‰¹æ®Šç¬¦å·ï¼‰
    df['æ¢¯æˆ·æ¯”ä¾‹'] = df['æ¢¯æˆ·æ¯”ä¾‹'].str.replace(r'[â–¡\s]', '', regex=True)
    
    # æ­£åˆ™è¡¨è¾¾å¼æå–æ•°å€¼
    pattern = r'(\d+)æ¢¯(\d+)æˆ·'
    extracted = df['æ¢¯æˆ·æ¯”ä¾‹'].str.extract(pattern)
    
    # è½¬æ¢ä¸ºæ•°å€¼å‹
    df['ç”µæ¢¯æ•°é‡'] = extracted[0].fillna('1').astype(int)  # é»˜è®¤1æ¢¯
    df['æ¯æ¢¯æˆ·æ•°'] = extracted[1].fillna('1').astype(int)  # é»˜è®¤1æˆ·
    
    # è®¡ç®—æ€»æˆ·æ•°å’ŒæœåŠ¡å¯†åº¦
    df['æ€»æˆ·æ•°'] = df['ç”µæ¢¯æ•°é‡'] * df['æ¯æ¢¯æˆ·æ•°']
    df['ç”µæ¢¯æœåŠ¡å¯†åº¦'] = df['æ€»æˆ·æ•°'] / df['ç”µæ¢¯æ•°é‡']
    
    # å¼‚å¸¸å€¼å¤„ç†ï¼ˆåŸºäºä¸šåŠ¡é€»è¾‘ï¼‰
    df.loc[df['ç”µæ¢¯æ•°é‡'] > 20, 'ç”µæ¢¯æ•°é‡'] = 20
    df.loc[df['æ¯æ¢¯æˆ·æ•°'] > 100, 'æ¯æ¢¯æˆ·æ•°'] = 100
    
    # åˆ›å»ºè¡ç”Ÿç‰¹å¾ï¼ˆä¿®æ­£åˆ—åå‰ç¼€ï¼‰
    df['æ‹¥æŒ¤åº¦'] = pd.cut(df['ç”µæ¢¯æœåŠ¡å¯†åº¦'],
                       bins=[0, 4, 8, 12, 20, 1000],
                       labels=['éå¸¸å®½æ¾', 'èˆ’é€‚', 'ä¸€èˆ¬', 'æ‹¥æŒ¤', 'éå¸¸æ‹¥æŒ¤'])
    df['è±ªå®…é…ç½®'] = (df['ç”µæ¢¯æœåŠ¡å¯†åº¦'] <= 4).astype(int)
    df['ç»æµå‹'] = (df['ç”µæ¢¯æœåŠ¡å¯†åº¦'] >= 12).astype(int)
    
    return df

# ==================== è®­ç»ƒé›†å¤„ç† ====================
# åŠ è½½ä¹‹å‰åˆå¹¶çš„æ•°æ®
train_merged_path = f"{output_dir}/train_merged_v6.csv"
train_df = pd.read_csv(train_merged_path)

# åº”ç”¨æ¢¯æˆ·æ¯”ä¾‹é¢„å¤„ç†
train_elevator_processed = preprocess_elevator_ratio(train_df)

# éªŒè¯å¤„ç†ç»“æœ
print("===== æ¢¯æˆ·æ¯”ä¾‹å¤„ç†éªŒè¯ =====")
print("1. ç”µæ¢¯æ•°é‡åˆ†å¸ƒ:")
print(train_elevator_processed['ç”µæ¢¯æ•°é‡'].value_counts().sort_index())

print("\n2. æ¯æ¢¯æˆ·æ•°æè¿°ç»Ÿè®¡:")
print(train_elevator_processed['æ¯æ¢¯æˆ·æ•°'].describe())

print("\n3. æ‹¥æŒ¤åº¦åˆ†å¸ƒ:")
print(train_elevator_processed['æ‹¥æŒ¤åº¦'].value_counts())

# ä¿å­˜æ¢¯æˆ·ç‰¹å¾ï¼ˆä¿®æ­£åˆ—åå‰ç¼€é€»è¾‘ï¼‰
elevator_cols = ['ç”µæ¢¯æ•°é‡', 'æ¯æ¢¯æˆ·æ•°', 'æ€»æˆ·æ•°', 'ç”µæ¢¯æœåŠ¡å¯†åº¦', 'æ‹¥æŒ¤åº¦', 'è±ªå®…é…ç½®', 'ç»æµå‹']
train_elevator_features = train_elevator_processed[elevator_cols]

# æ·»åŠ å‰ç¼€ï¼ˆç¡®ä¿ä¸åç»­å¼•ç”¨ä¸€è‡´ï¼‰
train_elevator_features.columns = ['æ¢¯æˆ·_' + col for col in train_elevator_features.columns]

# ä¿å­˜åˆ°columnsç›®å½•
train_elevator_path = f"{output_dir}/columns/train_æ¢¯æˆ·æ¯”ä¾‹.csv"
train_elevator_features.to_csv(train_elevator_path, index=False, encoding='utf_8_sig')
print(f"\nğŸ’¾ æ¢¯æˆ·ç‰¹å¾å·²ä¿å­˜åˆ°: {train_elevator_path}")

# æ›´æ–°åˆå¹¶æ–‡ä»¶ï¼ˆä¿®æ­£åˆ—åå¼•ç”¨ï¼‰
def update_train_merged():
    merged_train = pd.read_csv(train_merged_path)
    elevator_features = pd.read_csv(train_elevator_path)
    final_train = pd.concat([merged_train, elevator_features], axis=1)
    
    final_path = f"{output_dir}/train_merged_v7.csv"
    final_train.to_csv(final_path, index=False, encoding='utf_8_sig')
    print(f"ğŸ”„ åˆå¹¶åçš„è®­ç»ƒé›†å·²æ›´æ–°: {final_path}")
    
    # éªŒè¯åˆå¹¶åçš„åˆ—å
    print("\nåˆå¹¶åçš„æ–°å¢åˆ—:", [col for col in final_train.columns if col.startswith('æ¢¯æˆ·_')])
    return final_train

final_train_df = update_train_merged()

# æ˜¾ç¤ºå¤„ç†æ ·ä¾‹ï¼ˆä½¿ç”¨æ­£ç¡®çš„åˆ—åï¼‰
print("\n===== å¤„ç†ç»“æœæ ·ä¾‹ =====")
sample_cols = ['æ¢¯æˆ·æ¯”ä¾‹'] + [col for col in final_train_df.columns if col.startswith('æ¢¯æˆ·_')][:3]
display(final_train_df[sample_cols].head(5))
```

```python
# åŠ è½½æµ‹è¯•é›†åˆå¹¶æ•°æ®
test_merged_path = f"{output_dir}/test_merged_v6.csv"
test_df = pd.read_csv(test_merged_path)

def preprocess_test_elevator_ratio(df):
    """
    æµ‹è¯•é›†æ¢¯æˆ·æ¯”ä¾‹é¢„å¤„ç†ï¼ˆä¸è®­ç»ƒé›†å®Œå…¨ä¸€è‡´ï¼‰
    å‚æ•°:
        df: æµ‹è¯•é›†DataFrame
    è¿”å›:
        å¤„ç†åçš„ç‰¹å¾DataFrame
    """
    df = df.copy()
    
    # ç»Ÿä¸€æ ¼å¼å¤„ç†ï¼ˆä¸è®­ç»ƒé›†ç›¸åŒï¼‰
    df['æ¢¯æˆ·æ¯”ä¾‹'] = df['æ¢¯æˆ·æ¯”ä¾‹'].str.replace(r'[â–¡\s]', '', regex=True)
    
    # ç›¸åŒçš„æ­£åˆ™è¡¨è¾¾å¼æå–
    pattern = r'(\d+)æ¢¯(\d+)æˆ·'
    extracted = df['æ¢¯æˆ·æ¯”ä¾‹'].str.extract(pattern)
    
    # ç›¸åŒçš„é»˜è®¤å€¼å¤„ç†
    df['ç”µæ¢¯æ•°é‡'] = extracted[0].fillna('1').astype(int)
    df['æ¯æ¢¯æˆ·æ•°'] = extracted[1].fillna('1').astype(int)
    
    # ç›¸åŒçš„è®¡ç®—é€»è¾‘
    df['æ€»æˆ·æ•°'] = df['ç”µæ¢¯æ•°é‡'] * df['æ¯æ¢¯æˆ·æ•°']
    df['ç”µæ¢¯æœåŠ¡å¯†åº¦'] = df['æ€»æˆ·æ•°'] / df['ç”µæ¢¯æ•°é‡']
    
    # ç›¸åŒçš„å¼‚å¸¸å€¼é˜ˆå€¼
    df.loc[df['ç”µæ¢¯æ•°é‡'] > 20, 'ç”µæ¢¯æ•°é‡'] = 20
    df.loc[df['æ¯æ¢¯æˆ·æ•°'] > 100, 'æ¯æ¢¯æˆ·æ•°'] = 100
    
    # ç›¸åŒçš„åˆ†ç®±è¾¹ç•Œ
    bins = [0, 4, 8, 12, 20, 1000]
    labels = ['éå¸¸å®½æ¾', 'èˆ’é€‚', 'ä¸€èˆ¬', 'æ‹¥æŒ¤', 'éå¸¸æ‹¥æŒ¤']
    df['æ‹¥æŒ¤åº¦'] = pd.cut(df['ç”µæ¢¯æœåŠ¡å¯†åº¦'], bins=bins, labels=labels)
    
    # ç›¸åŒçš„è¡ç”Ÿç‰¹å¾
    df['è±ªå®…é…ç½®'] = (df['ç”µæ¢¯æœåŠ¡å¯†åº¦'] <= 4).astype(int)
    df['ç»æµå‹'] = (df['ç”µæ¢¯æœåŠ¡å¯†åº¦'] >= 12).astype(int)
    
    # ç›¸åŒçš„åˆ—åå‰ç¼€å¤„ç†
    features = df[['ç”µæ¢¯æ•°é‡', 'æ¯æ¢¯æˆ·æ•°', 'æ€»æˆ·æ•°', 'ç”µæ¢¯æœåŠ¡å¯†åº¦', 'æ‹¥æŒ¤åº¦', 'è±ªå®…é…ç½®', 'ç»æµå‹']]
    features.columns = ['æ¢¯æˆ·_' + col for col in features.columns]
    
    return features

# åº”ç”¨é¢„å¤„ç†
test_elevator_features = preprocess_test_elevator_ratio(test_df)

# éªŒè¯å¤„ç†ä¸€è‡´æ€§
print("===== æµ‹è¯•é›†æ¢¯æˆ·å¤„ç†éªŒè¯ =====")
print("1. ç”µæ¢¯æ•°é‡åˆ†å¸ƒ:")
print(test_elevator_features['æ¢¯æˆ·_ç”µæ¢¯æ•°é‡'].value_counts().sort_index())

print("\n2. æœåŠ¡å¯†åº¦åˆ†å¸ƒ:")
print(test_elevator_features['æ¢¯æˆ·_ç”µæ¢¯æœåŠ¡å¯†åº¦'].describe())

print("\n3. ä¸è®­ç»ƒé›†åˆ†ç®±å¯¹æ¯”:")
train_bins = pd.read_csv(f"{output_dir}/columns/train_æ¢¯æˆ·æ¯”ä¾‹.csv")
print("è®­ç»ƒé›†æ‹¥æŒ¤åº¦åˆ†å¸ƒ:\n", train_bins['æ¢¯æˆ·_æ‹¥æŒ¤åº¦'].value_counts())
print("\næµ‹è¯•é›†æ‹¥æŒ¤åº¦åˆ†å¸ƒ:\n", test_elevator_features['æ¢¯æˆ·_æ‹¥æŒ¤åº¦'].value_counts())

# ä¿å­˜åˆ°columnsç›®å½•
test_elevator_path = f"{output_dir}/columns/test_æ¢¯æˆ·æ¯”ä¾‹.csv"
test_elevator_features.to_csv(test_elevator_path, index=False, encoding='utf_8_sig')
print(f"\nğŸ’¾ æµ‹è¯•é›†æ¢¯æˆ·ç‰¹å¾å·²ä¿å­˜åˆ°: {test_elevator_path}")

# æ›´æ–°æµ‹è¯•é›†åˆå¹¶æ–‡ä»¶
def update_test_merged():
    merged_test = pd.read_csv(test_merged_path)
    elevator_features = pd.read_csv(test_elevator_path)
    final_test = pd.concat([merged_test, elevator_features], axis=1)
    
    final_path = f"{output_dir}/test_merged_v7.csv"
    final_test.to_csv(final_path, index=False, encoding='utf_8_sig')
    print(f"ğŸ”„ åˆå¹¶åçš„æµ‹è¯•é›†å·²æ›´æ–°: {final_path}")
    
    # åˆ—åéªŒè¯
    expected_cols = ['æ¢¯æˆ·_ç”µæ¢¯æ•°é‡', 'æ¢¯æˆ·_æ¯æ¢¯æˆ·æ•°', 'æ¢¯æˆ·_æ€»æˆ·æ•°', 
                    'æ¢¯æˆ·_ç”µæ¢¯æœåŠ¡å¯†åº¦', 'æ¢¯æˆ·_æ‹¥æŒ¤åº¦', 'æ¢¯æˆ·_è±ªå®…é…ç½®', 'æ¢¯æˆ·_ç»æµå‹']
    missing = set(expected_cols) - set(final_test.columns)
    assert len(missing) == 0, f"ç¼ºå¤±åˆ—: {missing}"
    
    return final_test

final_test_df = update_test_merged()

# æ˜¾ç¤ºå¤„ç†æ ·ä¾‹
print("\n===== æµ‹è¯•é›†å¤„ç†ç»“æœæ ·ä¾‹ =====")
sample_cols = ['æ¢¯æˆ·æ¯”ä¾‹'] + list(test_elevator_features.columns)[:4]
display(final_test_df[sample_cols].head(5))
```

```python
def preprocess_elevator(df):
    """
    é…å¤‡ç”µæ¢¯åˆ—é¢„å¤„ç†å‡½æ•°
    å¤„ç†é€»è¾‘ï¼š
    1. å¤„ç†ç©ºç™½å€¼ï¼ˆæ ¹æ®å»ºç­‘é«˜åº¦æ¨æ–­ï¼‰
    2. äºŒå…ƒç¼–ç ï¼ˆæœ‰=1ï¼Œæ— =0ï¼‰
    3. åˆ›å»ºè¡ç”Ÿç‰¹å¾
    """
    df = df.copy()
    
    # ç©ºç™½å€¼å¤„ç†è§„åˆ™ï¼ˆæ ¹æ®å»ºç­‘é«˜åº¦æ¨æ–­ï¼‰
    df['é…å¤‡ç”µæ¢¯'] = df['é…å¤‡ç”µæ¢¯'].fillna('æœªçŸ¥')
    df.loc[(df['é…å¤‡ç”µæ¢¯'] == 'æœªçŸ¥') & (df['æ¥¼å±‚_æ€»æ¥¼å±‚æ•°'] >= 7), 'é…å¤‡ç”µæ¢¯'] = 'æœ‰'  # 7å±‚ä»¥ä¸Šå»ºç­‘é»˜è®¤æœ‰ç”µæ¢¯
    df.loc[(df['é…å¤‡ç”µæ¢¯'] == 'æœªçŸ¥') & (df['æ¥¼å±‚_æ€»æ¥¼å±‚æ•°'] < 7), 'é…å¤‡ç”µæ¢¯'] = 'æ— '  # 7å±‚ä»¥ä¸‹é»˜è®¤æ— 
    
    # äºŒå…ƒç¼–ç 
    df['ç”µæ¢¯_é…å¤‡'] = df['é…å¤‡ç”µæ¢¯'].apply(lambda x: 1 if x == 'æœ‰' else 0)
    
    # åˆ›å»ºçŸ›ç›¾æ£€æµ‹ç‰¹å¾ï¼ˆå½“ä½æ¥¼å±‚æ ‡æ³¨æœ‰ç”µæ¢¯æ—¶æ ‡è®°ï¼‰
    if 'æ¥¼å±‚_æ€»æ¥¼å±‚æ•°' in df.columns:
        df['ç”µæ¢¯_çŸ›ç›¾'] = ((df['ç”µæ¢¯_é…å¤‡'] == 1) & (df['æ¥¼å±‚_æ€»æ¥¼å±‚æ•°'] < 3)).astype(int)
    
    return df

# ==================== è®­ç»ƒé›†å¤„ç† ====================
# åŠ è½½ä¹‹å‰åˆå¹¶çš„æ•°æ®ï¼ˆç¡®ä¿åŒ…å«æ¥¼å±‚ä¿¡æ¯ï¼‰
train_merged_path = f"{output_dir}/train_merged_v7.csv"
train_df = pd.read_csv(train_merged_path)

# åº”ç”¨ç”µæ¢¯é¢„å¤„ç†
train_elevator_processed = preprocess_elevator(train_df)

# éªŒè¯å¤„ç†ç»“æœ
print("===== ç”µæ¢¯é…å¤‡å¤„ç†éªŒè¯ =====")
print("1. åŸå§‹åˆ†å¸ƒ:")
print(train_elevator_processed['é…å¤‡ç”µæ¢¯'].value_counts(dropna=False))

print("\n2. ç¼–ç ååˆ†å¸ƒ:")
print("æœ‰ç”µæ¢¯å æ¯”:", f"{train_elevator_processed['ç”µæ¢¯_é…å¤‡'].mean():.1%}")

if 'ç”µæ¢¯_çŸ›ç›¾' in train_elevator_processed.columns:
    print("\n3. çŸ›ç›¾æ ‡è®°ï¼ˆä½å±‚æœ‰ç”µæ¢¯ï¼‰:", 
          f"{train_elevator_processed['ç”µæ¢¯_çŸ›ç›¾'].sum()}å¤„")

# ä¿å­˜ç”µæ¢¯ç‰¹å¾
elevator_cols = [col for col in train_elevator_processed.columns 
                if col.startswith('ç”µæ¢¯_')]
train_elevator_features = train_elevator_processed[elevator_cols]

# ä¿å­˜åˆ°columnsç›®å½•
train_elevator_path = f"{output_dir}/columns/train_é…å¤‡ç”µæ¢¯.csv"
train_elevator_features.to_csv(train_elevator_path, index=False, encoding='utf_8_sig')
print(f"\nğŸ’¾ ç”µæ¢¯ç‰¹å¾å·²ä¿å­˜åˆ°: {train_elevator_path}")

# æ›´æ–°åˆå¹¶æ–‡ä»¶
def update_train_merged():
    merged_train = pd.read_csv(train_merged_path)
    elevator_features = pd.read_csv(train_elevator_path)
    final_train = pd.concat([merged_train, elevator_features], axis=1)
    
    final_path = f"{output_dir}/train_merged_v8.csv"
    final_train.to_csv(final_path, index=False, encoding='utf_8_sig')
    print(f"ğŸ”„ åˆå¹¶åçš„è®­ç»ƒé›†å·²æ›´æ–°: {final_path}")
    return final_train

final_train_df = update_train_merged()

# æ˜¾ç¤ºå¤„ç†æ ·ä¾‹
print("\n===== å¤„ç†ç»“æœæ ·ä¾‹ =====")
sample_cols = ['é…å¤‡ç”µæ¢¯', 'ç”µæ¢¯_é…å¤‡']
if 'ç”µæ¢¯_çŸ›ç›¾' in final_train_df.columns:
    sample_cols.append('ç”µæ¢¯_çŸ›ç›¾')
display(final_train_df[sample_cols].head(5))
```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split

# åŠ è½½æœ€æ–°è®­ç»ƒé›†
train_path = f"{output_dir}/train_merged_v7.csv"
train_df = pd.read_csv(train_path)

def preprocess_remaining_columns(df):
    """å¤„ç†å‰©ä½™æ‰€æœ‰åˆ—çš„ç»Ÿä¸€å‡½æ•°"""
    df = df.copy()
    
    # 1. æ—¶é—´åˆ—å¤„ç†ï¼ˆç»Ÿä¸€æ ¼å¼åæå–å¹´ä»½ï¼‰
    date_cols = {
        'äº¤æ˜“æ—¶é—´': 'äº¤æ˜“å¹´ä»½',
        'ä¸Šæ¬¡äº¤æ˜“': 'ä¸Šæ¬¡äº¤æ˜“å¹´ä»½'
    }
    
    for col, new_col in date_cols.items():
        if col in df.columns:
            # ç»Ÿä¸€æ—¥æœŸæ ¼å¼ï¼ˆå¤„ç†éå¸¸è§„æ ¼å¼ï¼‰
            df[col] = pd.to_datetime(
                df[col].astype(str).str.replace(r'[^0-9/-]', '', regex=True),
                errors='coerce'
            )
            # æå–å¹´ä»½å¹¶å¤„ç†å¼‚å¸¸å€¼
            df[new_col] = df[col].dt.year
            df[new_col] = df[new_col].mask(df[new_col] > 2100, np.nan)
            df[new_col] = df[new_col].mask(df[new_col] < 1990, np.nan)
            # ç”¨ä¸­ä½æ•°å¡«è¡¥ç¼ºå¤±
            df[new_col] = df[new_col].fillna(df[new_col].median()).astype(int)
    
    # 2. æˆ¿å±‹å¹´é™å¤„ç†
    if 'æˆ¿å±‹å¹´é™' in df.columns:
        year_mapping = {
            'æœªæ»¡ä¸¤å¹´': 1,
            'æ»¡ä¸¤å¹´': 2, 
            'æ»¡äº”å¹´': 5
        }
        df['æˆ¿å±‹å¹´é™_ç¼–ç '] = df['æˆ¿å±‹å¹´é™'].map(year_mapping)
        df['æˆ¿å±‹å¹´é™_ç¼–ç '] = df['æˆ¿å±‹å¹´é™_ç¼–ç '].fillna(2)  # é»˜è®¤æ»¡ä¸¤å¹´
    
    # 3. åæ ‡å¤„ç†
    coord_cols = ['lon', 'lat']
    for col in coord_cols:
        if col in df.columns:
            df[col] = pd.to_numeric(df[col], errors='coerce')
            # ç”¨åŒå°åŒºåæ ‡ä¸­ä½æ•°å¡«è¡¥
            if 'å°åŒºåç§°' in df.columns:
                df[col] = df.groupby('å°åŒºåç§°')[col].transform(
                    lambda x: x.fillna(x.median()))
            df[col] = df[col].fillna(df[col].median())
    
    # 4. å¹´ä»½åˆ—å¤„ç†ï¼ˆéæ—¥æœŸï¼‰
    if 'å¹´ä»½' in df.columns:
        df['å¹´ä»½'] = pd.to_numeric(df['å¹´ä»½'], errors='coerce')
        df['å¹´ä»½'] = df['å¹´ä»½'].fillna(df['å¹´ä»½'].median()).astype(int)
    
    # 5. æ˜ç¡®è¦åˆ é™¤çš„åˆ—
    drop_cols = [
        'åˆ«å¢…ç±»å‹', 'äº¤æ˜“æƒå±', 'æˆ¿å±‹ç”¨é€”', 'äº§æƒæ‰€å±', 'æŠµæŠ¼ä¿¡æ¯',
        'æˆ¿å±‹ä¼˜åŠ¿', 'æ ¸å¿ƒå–ç‚¹', 'æˆ·å‹ä»‹ç»', 'å‘¨è¾¹é…å¥—', 'äº¤é€šå‡ºè¡Œ',
        'äº¤æ˜“æ—¶é—´', 'ä¸Šæ¬¡äº¤æ˜“', 'æˆ¿å±‹å¹´é™'  # åŸå§‹åˆ—ï¼ˆå·²æå–æ–°ç‰¹å¾ï¼‰
    ]
    df = df.drop(columns=[c for c in drop_cols if c in df.columns], errors='ignore')
    
    return df

# åº”ç”¨å¤„ç†
train_processed = preprocess_remaining_columns(train_df)
```

```python
def convert_to_numeric(df):
    """å°†éæ•°å€¼åˆ—ç³»ç»ŸåŒ–è½¬æ¢ä¸ºæ•°å€¼"""
    
    # 1. æœ‰åºç±»åˆ«åˆ—ç¼–ç 
    ordinal_cols = {
        'é¢ç§¯_é¢ç§¯ç­‰çº§': ['æå°', 'å°', 'ä¸­', 'å¤§', 'è¶…å¤§', 'è±ªå®…'],
        'æ¢¯æˆ·_æ‹¥æŒ¤åº¦': ['éå¸¸å®½æ¾', 'èˆ’é€‚', 'ä¸€èˆ¬', 'æ‹¥æŒ¤', 'éå¸¸æ‹¥æŒ¤'],
        'é…å¤‡ç”µæ¢¯': ['æ— ', 'æœ‰']
    }
    
    for col, categories in ordinal_cols.items():
        if col in df.columns:
            df[col] = pd.Categorical(df[col], categories=categories, ordered=True)
            df[f"{col}_ç¼–ç "] = df[col].cat.codes
    
    # 2. One-Hotç¼–ç é«˜åŸºæ•°åˆ—
    onehot_cols = [col for col in ['æœå‘_ä¸»æ–¹å‘', 'æ¥¼å±‚_æ¥¼å±‚ç±»å‹'] 
                  if col in df.columns and df[col].nunique() < 20]
    
    for col in onehot_cols:
        encoded = pd.get_dummies(df[col], prefix=col)
        df = pd.concat([df.drop(col, axis=1), encoded], axis=1)
    
    
    
    # 4. åˆ é™¤å·²æå–ç‰¹å¾çš„åŸå§‹åˆ—
    drop_cols = [
        'ç¯çº¿', 'æˆ¿å±‹æˆ·å‹', 'æ‰€åœ¨æ¥¼å±‚', 'å»ºç­‘ç»“æ„', 
        'è£…ä¿®æƒ…å†µ', 'æ¢¯æˆ·æ¯”ä¾‹', 'é¢ç§¯_é¢ç§¯ç­‰çº§',
        'æ¢¯æˆ·_æ‹¥æŒ¤åº¦', 'é…å¤‡ç”µæ¢¯'
    ]
    df = df.drop(columns=[c for c in drop_cols if c in df.columns])
    
    return df

# åº”ç”¨å¤„ç†
train_processed = convert_to_numeric(train_processed)
test_processed = convert_to_numeric(test_processed)

# éªŒè¯å¤„ç†ç»“æœ
print("è®­ç»ƒé›†å‰©ä½™éæ•°å€¼åˆ—:", train_processed.select_dtypes(exclude='number').columns.tolist())
print("æµ‹è¯•é›†å‰©ä½™éæ•°å€¼åˆ—:", test_processed.select_dtypes(exclude='number').columns.tolist())
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression, Lasso, Ridge, ElasticNet
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.model_selection import cross_val_score
import matplotlib.pyplot as plt

# åŠ è½½æ•°æ®
X_train = pd.read_csv("/home/mw/project/output/X_train.csv")
y_train = pd.read_csv("/home/mw/project/output/y_train.csv").squeeze()  # è½¬ä¸ºSeries
X_test = pd.read_csv("/home/mw/project/output/X_test.csv")
y_test = pd.read_csv("/home/mw/project/output/y_test.csv").squeeze()

```

```python
def train_and_evaluate_model(model, X_train, y_train, X_test, y_test, model_name):
    """è®­ç»ƒæ¨¡å‹å¹¶è®¡ç®—å„é¡¹æŒ‡æ ‡"""
    # è®­ç»ƒæ¨¡å‹
    model.fit(X_train, y_train)
    
    # é¢„æµ‹ç»“æœ
    y_train_pred = model.predict(X_train)
    y_test_pred = model.predict(X_test)
    
    # è®¡ç®—æŒ‡æ ‡
    metrics = {
        'In sample': {
            'MAE': mean_absolute_error(y_train, y_train_pred),
            'RMSE': np.sqrt(mean_squared_error(y_train, y_train_pred)),
            'R2': model.score(X_train, y_train)
        },
        'Out of sample': {
            'MAE': mean_absolute_error(y_test, y_test_pred),
            'RMSE': np.sqrt(mean_squared_error(y_test, y_test_pred)),
            'R2': model.score(X_test, y_test)
        }
    }
    
    # äº¤å‰éªŒè¯ (6æŠ˜)
    cv_mae = -cross_val_score(model, X_train, y_train, 
                             cv=6, scoring='neg_mean_absolute_error').mean()
    cv_rmse = np.sqrt(-cross_val_score(model, X_train, y_train, 
                                      cv=6, scoring='neg_mean_squared_error').mean())
    cv_r2 = cross_val_score(model, X_train, y_train, cv=6, scoring='r2').mean()
    
    metrics['Cross-validation'] = {
        'MAE': cv_mae,
        'RMSE': cv_rmse,
        'R2': cv_r2
    }
    
    # æ‰“å°ç»“æœ
    print(f"\n===== {model_name} æ¨¡å‹ç»“æœ =====")
    print(f"In-sample R2: {metrics['In sample']['R2']:.4f}")
    print(f"Out-of-sample R2: {metrics['Out of sample']['R2']:.4f}")
    print(f"6æŠ˜äº¤å‰éªŒè¯ R2: {metrics['Cross-validation']['R2']:.4f}")
    
    return model, metrics

# åˆå§‹åŒ–çº¿æ€§å›å½’æ¨¡å‹
lr = LinearRegression()

# è®­ç»ƒå¹¶è¯„ä¼°
lr_model, lr_metrics = train_and_evaluate_model(
    lr, X_train, y_train, X_test, y_test, "Linear Regression"
)
```

```python
def plot_results(y_true, y_pred, title):
    """ç»˜åˆ¶å®é™…å€¼ vs é¢„æµ‹å€¼"""
    plt.figure(figsize=(10, 6))
    plt.scatter(y_true, y_pred, alpha=0.3)
    plt.plot([y_true.min(), y_true.max()], 
             [y_true.min(), y_true.max()], 'r--')
    plt.xlabel('Actual Price')
    plt.ylabel('Predicted Price')
    plt.title(title)
    plt.grid(True)
    plt.show()

# ç»˜åˆ¶æµ‹è¯•é›†ç»“æœ
y_test_pred = lr_model.predict(X_test)
plot_results(y_test, y_test_pred, "Linear Regression - Test Set Performance")
```

```python
def create_metrics_table(metrics_dict, model_name):
    """åˆ›å»ºæŒ‡æ ‡è¡¨æ ¼"""
    df = pd.DataFrame.from_dict(metrics_dict, orient='index')
    df = df[['MAE', 'RMSE', 'R2']]
    df['Model'] = model_name
    return df

# åˆ›å»ºæŒ‡æ ‡è¡¨æ ¼
lr_metrics_df = create_metrics_table(lr_metrics, "Linear Regression")
print("\nçº¿æ€§å›å½’æ€§èƒ½æŒ‡æ ‡:")
display(lr_metrics_df)
```
