```python
import pandas as pd
import os
import re
import time
from docxtpl import DocxTemplate
```

```python
os.chdir("/Users/qiudanqi/homework2/")  # 切换到正确目录
os.listdir()
```

```python
# 1. 读取 Excel 文件
excel_file = "/Users/qiudanqi/homework2/schoollist.xlsx"  
df = pd.read_excel(excel_file)  
```

```python
# 2. 读取 Word 模板
template_file = "/Users/qiudanqi/homework2/personal_statement.docx"  # 你的 Word 模板
doc = DocxTemplate(template_file)  # 假设模板文件和脚本在同一目录

```

```python
# 3. 创建保存目录
output_folder = "Application_Letters"
os.makedirs(output_folder, exist_ok=True)  # 创建文件夹（如果不存在）
```

```python


# 4. 生成申请信
for index, row in df.iterrows():
    university = row["university"]
    for j in range(1, 4):  # 每所大学 3 个专业
        major = row[f"major{j}"]  # Excel 的专业列应命名为 Program1, Program2, Program3
        

        # 替换 Word 模板中的变量
        context = {"university": university, "major": major}
        doc.render(context)
        # 生成合法的文件名
        filename = f"{university}_{major}.docx"
        filepath = os.path.join(output_folder, filename)  

        # 保存 Word 文件
        doc.save(filepath)
        print(f"已生成：{filepath}")

        

print("所有申请信已生成完成！")
```
