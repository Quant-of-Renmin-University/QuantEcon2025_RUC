### 数据处理

```python
#绿化率：改变数据类型、去除超过100%的异常值（已知无其他异常值）、用中位数填充空的
xiaoqu['绿 化 率'] = xiaoqu['绿 化 率'].apply(lambda x: float(x[:-1]) if pd.notna(x) else x)
xiaoqu['绿 化 率'] = xiaoqu['绿 化 率'].where((xiaoqu['绿 化 率'].isnull()) | (xiaoqu['绿 化 率'] <= 100), None)
xiaoqu['绿 化 率'].fillna(xiaoqu['绿 化 率'].median(), inplace=True)

#容积率：用中位数填充
xiaoqu['容 积 率'].fillna(xiaoqu['容 积 率'].median(), inplace=True)

#房屋总数：转换类型
xiaoqu['房屋总数'] = xiaoqu['房屋总数'].apply(lambda x: int(x[:-1]) if pd.notna(x) else x)

#楼栋总数：转换类型
xiaoqu['楼栋总数'] = xiaoqu['楼栋总数'].apply(lambda x: int(x[:-1]) if pd.notna(x) else x)
```

停车问题

```python
filtered_df = xiaoqu[xiaoqu['停车位'] > 5000]

# 打印筛选后的结果
print(filtered_df[['物业类别','房屋总数','停车位']])
```

```python
#结合上述信息，把几个拥有太多停车位的普通住宅按照一户两车进行填充
xiaoqu.loc[[871,1973,2551,2552], '停车位'] = xiaoqu.loc[[871,1973,2551,2552], '房屋总数']*2
#把nan填充为0
xiaoqu['停车位'] = xiaoqu['停车位'].fillna(0)
xiaoqu['停车位'].describe()
```

```python
#处理建筑年代
def process_cell(cell):
    if pd.isna(cell):
        return cell
    two_years = re.findall(r'\d+', cell)
    if len(two_years) == 2:
        year1, year2 = map(int, two_years)

        #把2024年以后的年份视为异常值处理。
        if year1 > 2024 and year2 > 2024:
            return None
        elif year1 > 2024:
            return float(year2)
        elif year2 > 2024:
            return float(year1)
        return (year1 + year2) / 2
    elif len(two_years) == 1:
        year = int(two_years[0])

        #把2024年以后的年份视为异常值处理。
        if year > 2024:
            return None
        return float(year)
    return cell

xiaoqu['建筑年代'] = xiaoqu['建筑年代'].apply(process_cell)
#用中位数填充空值
xiaoqu['建筑年代'].fillna(xiaoqu['建筑年代'].median(), inplace=True)
```

```python
#丢弃不需要的列
xiaoqu.drop(['区县', '城市','板块','小区地址','物业类别','开发商','建筑结构','物业办公电话','产权描述'], axis=1, inplace=True)
```

```python
#停车费用处理
new_data = []
for fee in xiaoqu['停车费用']:
    if pd.isnull(fee) or fee == '0' or fee == '0.0' or fee == '0.00' or fee == '0.000000000000000000000' or fee == '暂无' or fee == '无' or fee == '免费' or fee == '-':
        new_data.append(0)
    elif isinstance(fee, str):
        nums = [float(s) for s in fee.split() if s.isdigit() or s.replace('.', '', 1).isdigit()]
        if nums:
            new_data.append(np.mean(nums))
        else:
            new_data.append(0)
    else:
        new_data.append(float(fee))

# 将处理后的数据替换原列数据
xiaoqu['停车费用'] = new_data
```
