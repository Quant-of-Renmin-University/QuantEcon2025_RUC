```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 读取数据
df = pd.read_csv('一次分类数据/output_2.csv', encoding='utf-8')
# 使用 seaborn 热力图
plt.figure(figsize=(10, 6))

# 绘制热力图
sns.kdeplot(x=df['lon'], y=df['lat'], cmap="coolwarm", fill=True, thresh=0, levels=100, alpha=0.6)
# 计算单价
df['单价'] = df['价格'] / df['建筑面积']
# 添加价格单价的颜色映射
plt.scatter(df['lon'], df['lat'], c=df['单价'], cmap='coolwarm', alpha=0.5)

plt.title("房价单价的空间分布热力图")
plt.xlabel("经度 (Longitude)")
plt.ylabel("纬度 (Latitude)")
plt.colorbar(label="单价 (元/平米)")
plt.show()
plt.savefig('热力图0.png')
```
