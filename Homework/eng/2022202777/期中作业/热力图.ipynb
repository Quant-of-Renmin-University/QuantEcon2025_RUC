```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# 
df = pd.read_csv('/output_2.csv', encoding='utf-8')
#  seaborn 
plt.figure(figsize=(10, 6))

# 
sns.kdeplot(x=df['lon'], y=df['lat'], cmap="coolwarm", fill=True, thresh=0, levels=100, alpha=0.6)
# 
df[''] = df[''] / df['']
# 
plt.scatter(df['lon'], df['lat'], c=df[''], cmap='coolwarm', alpha=0.5)

plt.title("")
plt.xlabel(" (Longitude)")
plt.ylabel(" (Latitude)")
plt.colorbar(label=" (/)")
plt.show()
plt.savefig('0.png')
```
