# 数据简单处理

# 建筑年代处理

```python
def parse_construction_year(s):
    """
    处理建筑年代字符串，返回建筑年份
    更新规则：
        1. 单个年份 → 直接提取
        2. 时间跨度 <=15年 → 取中间值
        3. 时间跨度 >15年:
           - 若起始年份 <1970 → 填补为1985
           - 若起始年份 >=1970 → 取起始年份
    输入示例: 
        - "1920-2016年" → 1985 (跨度96年>15且起始<1970)
        - "1975-2020年" → 1975 (跨度45年>15且起始>=1970)
        - "2005-2010年" → 2007 (跨度5年<=15)
        - "1995年" → 1995
    """
    try:
        s = str(s).strip()
        if not s or s.lower() in ['nan', '未知', '暂无', '无']:
            return np.nan
        
        # 匹配时间段格式（兼容中英文符号）
        range_match = re.search(r'(\d{3,4})\s*[-至到~]+\s*(\d{3,4})', s)
        if range_match:
            start = int(range_match.group(1))
            end = int(range_match.group(2))
            
            # 修正明显错误（如2015-2000）
            start, end = sorted([start, end])
            
            span = end - start
            
            # 长跨度特殊处理
            if span > 15:
                if start < 1970:
                    # 历史数据修正规则：1920s-1950s的异常数据统一设为1985
                    return 1985
                else:
                    # 现代长周期项目取起始年份（如1975-2020取1975）
                    return start
            else:
                # 短跨度取中间值
                return round((start + end) / 2)
        
        # 匹配单个年份（兼容带"年"字符）
        year_match = re.search(r'(\d{4})年?', s)
        if year_match:
            return int(year_match.group(1))
        
        # 处理模糊描述（如"2000年以前"取2000）
        if "以前" in s:
            year = re.search(r'\d{4}', s)
            return int(year.group()) if year else np.nan
            
        return np.nan
    except Exception as e:
        # 打印异常记录用于调试
        print(f"解析异常: {s} -> {str(e)}")
        return np.nan
```

```python
#应用函数
xiaoqu['建筑年代'] = xiaoqu['建筑年代'].apply(parse_construction_year)
```

```python
# 定义需要手动修正的小区及正确年份
YEAR_MAPPING = {
    "金台里": 1991,
    "人定湖西里": 1982,
    "三五四一社区": 1992
    # 可在此添加更多已知异常小区
}
```

```python
current_year = pd.Timestamp.now().year
xiaoqu.loc[xiaoqu['建筑年代'] > current_year, '建筑年代'] = np.nan
```

```python
for community, year in YEAR_MAPPING.items():
        mask = xiaoqu["名称"] == community
        if mask.any():
            xiaoqu.loc[mask, "建筑年代"] = year
```

```python
#停车费用处理
new_data = []
for fee in xiaoqu['停车费用']:
    if pd.isnull(fee) or fee == '0' or fee == '0.0' or fee == '0.00' or fee == '0.000000000000000000000' or fee == '暂无' or fee == '无' or fee == '免费' or fee == '-':
        new_data.append(0)
    elif isinstance(fee, str):
        nums = [float(s) for s in fee.split() if s.isdigit() or s.replace('.', '', 1).isdigit()]
        if nums:
            new_data.append(np.mean(nums))
        else:
            new_data.append(0)
    else:
        new_data.append(float(fee))

# 将处理后的数据替换原列数据
xiaoqu['停车费用'] = new_data
```
