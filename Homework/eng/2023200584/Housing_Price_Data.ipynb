```python
from selenium import webdriver  
import pandas as pd
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.common import NoSuchElementException,StaleElementReferenceException
from selenium.common.exceptions import NoSuchElementException, ElementClickInterceptedException
```

```python
# open the website

driver = webdriver.Edge()

url = "https://esf.fang.com/house-a015277-b02315/"
driver.get(url)
```

```python
string_list = []

i = 0

while i < 50:
    table =driver.find_element(By.CLASS_NAME,'shop_list')
    rows = table.find_elements(By.TAG_NAME, 'dl')  
    
    data = []
    for row in rows:    # 每一条dl标签中
        cells = row.find_elements(By.TAG_NAME, 'dd')
        if cells:    # 找到dd标签
            for cell in cells:    # 每一条dd标签
                anothor_inf = cell.find_elements(By.TAG_NAME, 'p') 
                price = cell.find_elements(By.TAG_NAME, 'span')
                if anothor_inf:    # 找到p标签
                    another_data = [another_inf1.text for another_inf1 in anothor_inf]
            
                elif price:    # 找到price标签
                    price_data = [price1.text for price1 in price]
                    another_data.append(price_data)
                    data.append(another_data)
    df = pd.DataFrame(data)
    string_list.append(df)
    print(i+1)    # 展示爬取进度
    
    # 检查是否存在下一页
    next_page_elements = driver.find_elements(By.XPATH, "//a[contains(text(), '下一页')]")
    if not next_page_elements:
        print('No more pages')
        break
    # 如果存在，打开下一页
    try:
        next_page = WebDriverWait(driver,10).until(EC.element_to_be_clickable((By.XPATH,  "//a[contains(text(),'下一页')]")))
        next_page_href = next_page.get_attribute('href')
        driver.get(next_page_href)
        i+=1
    except NoSuchElementException:
        print('No more pages')
    except ElementClickInterceptedException:
        print("not clickable")

        

driver.quit()
```

```python
df_full = pd.concat(string_list, axis = 0, ignore_index= True)
df_full
```

```python
# rearrange the data

#split the columns
df_full_split0 = df_full.iloc[:,0].str.split("|",expand=True) 
df_full_split1 = df_full.iloc[:,1].str.split("\n",expand=True)
df_full_split2_1 = df_full.iloc[:,2].str.split("\n").str[-1]
df_full_split2_2 = df_full.iloc[:,2].str.split("\n").str[:-1]
df_full.iloc[:,3] = df_full.iloc[:,3].astype(str)
df_full_split3 = df_full.iloc[:,3].str.split(",",expand=True)
# 合并并且重命名列
df_final = pd.concat([df_full_split0,df_full_split1,df_full_split2_1,df_full_split2_2,df_full_split3],axis=1)
df_final.columns = ['房型', '面积', '楼层', '朝向', '建成年份', '户主', '小区', '地址', '交通', '其他特征', '总售价', '每平方米售价'] 
# 更改数据类型并添加单位
df_final['总售价'] = df_final['总售价'].str.extract('(\\d+)')
df_final['总售价'] = df_final['总售价'] + '万'
df_final['每平方米售价'] = df_final['每平方米售价'].str.extract('(\\d+)')
df_final['每平方米售价'] = df_final['每平方米售价'] + '元/平方米'
# reorder columns
new_columns_order = ['小区', '地址', '房型', '面积', '楼层', '朝向', '建成年份', '户主', '交通', '其他特征', '总售价', '每平方米售价']
df_final = df_final[new_columns_order]

df_final

```
