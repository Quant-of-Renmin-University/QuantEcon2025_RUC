```python
import pandas as pd

filename='eng.xlsx'

df=pd.read_excel(filename,header=2)

cond=df['序号'].notnull()
df=df[cond]

df=df.astype({'学号':int})

var_list=['序号', '学号', '姓名', '班级']

df=df[var_list]
```

```python
df['班级'].replace('\d\d\d\d|级|\d|班','',regex=True).value_counts()
```

```python
df_groups=df.sample(frac = 1,random_state=202509)

df_groups['number'] = range(len(df_groups))

n_members=4

df_groups['group']=df_groups['number']//n_members+1

```

```python
df_groups.sort_values(by='group')
```
